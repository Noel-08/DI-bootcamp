## 🌟 Exercise 2: LLM Architecture and Application Scenarios

### BERT
- **Architectural Differences**: Bidirectional encoder-only; Masked Language Model (MLM) for pre-training, which predicts masked tokens based on both left and right context. Historically also used Next Sentence Prediction (NSP), though often omitted now.
- **Application**: Classification de texte (ex: analyse de sentiment, détection de spam), Reconnaissance d'Entités Nommées (NER), Réponse aux questions (Q&A).
- **Why**: Sa nature bidirectionnelle lui permet de comprendre le contexte complet d'un mot ou d'une phrase, ce qui est crucial pour les tâches où la compréhension fine du sens est requise (ex: savoir si un mot est un nom de personne ou de lieu).

### GPT
- **Architectural Differences**: Decoder-only; autoregressive causal language model. It predicts the next token in a sequence based on the preceding tokens, generating text unidirectionnellement.
- **Application**: Génération de texte (ex: écriture créative, résumé, complétion de code, chatbots conversationnels).
- **Why**: Son architecture de décodeur unidirectionnel et sa capacité à générer du texte de manière cohérente en prédisant le mot suivant le rendent idéal pour des tâches de génération où la fluidité et la créativité sont importantes.

### T5
- **Architectural Differences**: Encoder-decoder; text-to-text framework. It treats every NLP problem as a text-to-text task, where the input is text and the output is also text. It uses a standard Transformer architecture.
- **Application**: Traduction automatique, résumé de texte, simplification de texte, réponse à des questions, réécriture de phrases.
- **Why**: Sa structure encodeur-décodeur et son cadre unifié "texte-à-texte" le rendent extrêmement polyvalent. Il peut gérer une grande variété de tâches en les formulant comme des problèmes de transformation de texte à texte, ce qui permet une grande flexibilité et de bonnes performances sur de nombreuses tâches.

## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- Benefits (explain each in your words):
  1) **Improved generalization**: La pré-formation permet aux modèles d'apprendre des motifs et des structures linguistiques générales à partir de vastes corpus de données, ce qui les rend plus aptes à s'adapter et à performer sur une grande variété de tâches spécifiques, même sur des données qu'ils n'ont jamais vues explicitement pendant l'entraînement.
  2) **Less labeled data**: Grâce à la connaissance acquise lors de la pré-formation, les modèles nécessitent beaucoup moins de données étiquetées pour le fine-tuning sur une tâche spécifique. Cela réduit considérablement le coût et l'effort d'annotation des données.
  3) **Faster fine-tuning**: Le fine-tuning d'un modèle pré-entraîné est généralement beaucoup plus rapide que l'entraînement d'un modèle à partir de zéro. Le modèle a déjà une compréhension fondamentale du langage, et le fine-tuning se concentre uniquement sur l'adaptation à la spécificité de la tâche.
  4) **Transfer learning**: La pré-formation incarne le principe de l'apprentissage par transfert, où les connaissances acquises sur une tâche (la modélisation du langage) sont transférées et appliquées à d'autres tâches connexes, améliorant ainsi les performances et l'efficacité de l'entraînement.
  5) **Robustness**: Les modèles pré-entraînés sont souvent plus robustes face aux variations et au bruit dans les données. Ayant été exposés à une immense diversité de textes, ils développent une meilleure capacité à gérer des entrées imparfaites ou inattendues.

- Ethical concerns:
  - **Bias**: Les LLM sont entraînés sur d'énormes quantités de texte provenant d'Internet, qui contiennent inévitablement des biais sociaux (stéréotypes de genre, raciaux, etc.). Ces biais peuvent être amplifiés et perpétués par le modèle, conduisant à des prédictions ou des générations discriminatoires.
  - **Misinformation/Disinformation**: Les modèles peuvent générer des informations fausses ou trompeuses de manière très convaincante, ce qui peut contribuer à la propagation de la désinformation et avoir des impacts sociétaux négatifs.
  - **Misuse**: Les capacités de génération de texte des LLM peuvent être exploitées à des fins malveillantes, telles que la création de campagnes de phishing sophistiquées, la production de faux contenus pour manipuler l'opinion publique ou la rédaction de code malveillant.
  - **Privacy**: Les données d'entraînement peuvent contenir des informations personnelles identifiables (PII) ou sensibles. Bien que des efforts soient faits pour anonymiser les données, il existe un risque que les modèles mémorisent et régurgitent ces informations, compromettant la vie privée.

- Mitigations:
  - **Data curation**: Nettoyer et filtrer les ensembles de données d'entraînement pour réduire les biais toxiques et la désinformation. Cela inclut la déduplication, la suppression des contenus haineux, et l'équilibrage des représentations.
  - **Differential privacy**: Intégrer des techniques de confidentialité différentielle pendant l'entraînement pour garantir que la présence ou l'absence d'un point de donnée individuel dans l'ensemble d'entraînement n'affecte pas significativement le résultat du modèle, protégeant ainsi la vie privée.
  - **Safety filters**: Développer et intégrer des filtres post-génération ou des couches de modération pour détecter et bloquer les contenus générés qui sont nocifs, biaisés ou trompeurs.
  - **RLHF (Reinforcement Learning from Human Feedback)**: Utiliser le feedback humain pour aligner le comportement des modèles sur des valeurs et des normes éthiques, en les incitant à générer des réponses plus utiles, honnêtes et inoffensives.
  - **Audits and transparency**: Effectuer des audits réguliers des modèles pour identifier et quantifier les biais, et rendre les processus de développement et d'entraînement plus transparents pour la recherche et l'examen public. ## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- **Describe the Q/K/V flow, softmax weighting, and value mixing.**
  Dans le mécanisme d'auto-attention, pour chaque mot (token) de la séquence d'entrée, trois vecteurs sont générés : Query (Q), Key (K) et Value (V). Ces vecteurs sont obtenus en multipliant l'embedding du mot par trois matrices de poids distinctes. L'attention est calculée en prenant le produit scalaire entre le vecteur Query du mot actuel et tous les vecteurs Key de tous les mots de la séquence. Ce produit scalaire mesure la similarité (pertinence) entre le mot Query et chaque autre mot Key. Les scores de similarité sont ensuite normalisés par une fonction Softmax, ce qui produit des poids d'attention qui indiquent l'importance de chaque mot K par rapport à Q. Enfin, ces poids sont multipliés par les vecteurs Value correspondants, et la somme pondérée de ces vecteurs Value constitue la sortie de l'auto-attention pour le mot Q, représentant son nouveau contexte.

- **Explain why multiple heads help (subspace projections, diverse relations).**
  L'attention multi-têtes (Multi-Head Attention) améliore l'auto-attention en permettant au modèle de se concentrer sur différentes "facettes" ou "types de relations" simultanément. Au lieu de réaliser un seul calcul d'attention, le mécanisme divise les vecteurs Q, K et V en plusieurs "têtes" (heads) plus petites, chacune avec ses propres matrices de poids. Chaque tête peut alors apprendre à focaliser son attention sur un aspect différent de la séquence d'entrée (par exemple, une tête pourrait identifier les relations syntaxiques, une autre les relations sémantiques, une autre encore la coréférence). Les résultats de chaque tête sont ensuite concaténés et projetés linéairement pour former la sortie finale. Cela permet au modèle de capturer des relations plus riches et diverses dans les données, améliorant ainsi sa capacité à comprendre le contexte et la sémantique.

- **Example sentence (different from lesson):**
  - Sentence: "The quick brown fox jumps over the lazy dog."
  - **Relations captured by distinct heads**:
    - **Head 1 (relation sujet-verbe)**: Pour le mot "jumps", une tête pourrait donner un poids élevé à "fox" (le sujet) et un poids modéré à "quick" et "brown" (les adjectifs du sujet). Cela aide à comprendre qui fait l'action.
    - **Head 2 (relation objet-prépositionnel)**: Pour le mot "dog", une autre tête pourrait donner un poids élevé à "lazy" (l'adjectif du chien) et à "over" (la préposition qui le lie à l'action), pour comprendre l'objet de l'action et sa position relative.
    - **Head 3 (relation temporelle/adverbiale)**: Si la phrase était plus complexe, une tête pourrait se concentrer sur les relations temporelles ou adverbiales, par exemple, entre un verbe et un adverbe de temps.

### Pre-training Objectives
- **MLM vs CLM: compare objectives and typical use.**
  - **MLM (Masked Language Modeling)**: L'objectif est de prédire des mots masqués aléatoirement dans une phrase, en utilisant le contexte bidirectionnel (mots à gauche et à droite du mot masqué). Le modèle est entraîné à "remplir les blancs".
    - **Typical use**: Modèles de compréhension de langage comme BERT, conçus pour des tâches où une compréhension contextuelle profonde est essentielle, telles que la classification de texte, la réponse aux questions, la reconnaissance d'entités nommées.
  - **CLM (Causal Language Modeling)**: L'objectif est de prédire le mot suivant dans une séquence, en se basant uniquement sur les mots précédents (contexte unidirectionnel). Le modèle génère du texte de manière auto-régressive.
    - **Typical use**: Modèles de génération de langage comme GPT, conçus pour la génération de texte, la traduction, la complétion de code, les chatbots conversationnels, où la fluidité et la cohérence de la sortie sont primordiales.

- **When to prefer MLM vs CLM.**
  - **Préférer MLM** quand la tâche requiert une compréhension profonde et bidirectionnelle du contexte, par exemple pour analyser le sentiment d'un texte, extraire des informations précises, ou effectuer une classification. Le modèle a besoin de voir tout le contexte pour prendre une décision éclairée.
  - **Préférer CLM** quand l'objectif est de générer du texte de manière fluide et cohérente, par exemple pour écrire un article, répondre à une question ouverte, ou poursuivre une conversation. Le modèle doit produire du texte mot par mot, en se basant sur ce qu'il a déjà généré.

- **NSP: why early BERT used it and why many modern models avoid it.**
  - **Early BERT used NSP** pour aider le modèle à comprendre les relations entre les phrases, ce qui est crucial pour des tâches comme la réponse aux questions ou l'inférence en lecture. L'objectif était de prédire si deux segments de texte consécutifs étaient adjacents dans le document original (Next Sentence Prediction). Cela incitait le modèle à apprendre des représentations qui encodent non seulement le sens des mots, mais aussi la cohérence et le flux narratif entre les phrases.
  - **Modern models often avoid it** car il a été constaté que l'objectif NSP n'apportait pas toujours une amélioration significative des performances et pouvait même parfois nuire à certaines tâches. Les recherches ont montré que l'objectif MLM seul est souvent suffisant pour apprendre des représentations de phrases efficaces. De plus, la tâche NSP est parfois trop simple ou ne capture pas les relations inter-phrases les plus importantes pour les applications modernes, et sa suppression simplifie l'entraînement et réduit la complexité du modèle.

### Transformer Model Selection
- **Sentiment on reviews (Encoder-only / Decoder-only / Encoder-Decoder) + justification**
  - **Type**: **Encoder-only** (ex: BERT).
  - **Justification**: L'analyse de sentiment est une tâche de classification qui nécessite une compréhension profonde et bidirectionnelle du texte d'entrée pour capturer toutes les nuances émotionnelles. Un encodeur seul, comme BERT, excelle à créer des représentations contextuelles riches d'une phrase complète, ce qui est parfait pour classer un avis comme positif, négatif ou neutre. Il n'y a pas de texte à générer, seulement à comprendre.

- **Conversational chatbot (creative responses): (Encoder-only / Decoder-only / Encoder-Decoder) + justification**
  - **Type**: **Decoder-only** (ex: GPT).
  - **Justification**: Un chatbot qui génère des réponses créatives et engageantes nécessite la capacité de produire du texte de manière auto-régressive, un mot à la fois, en se basant sur la conversation précédente. L'architecture de décodeur seul est spécifiquement conçue pour la génération de langage, permettant une production de texte fluide, cohérente et contextuellement pertinente, idéale pour les interactions conversationnelles ouvertes.

- **Technical document translation (EN→ES): (Encoder-only / Decoder-only / Encoder-Decoder) + justification**
  - **Type**: **Encoder-Decoder** (ex: T5).
  - **Justification**: La traduction est une tâche classique de séquence à séquence. Un modèle encodeur-décodeur est idéal car l'encodeur peut traiter et comprendre l'intégralité du document technique source (en anglais), construisant une représentation sémantique riche. Le décodeur utilise ensuite cette représentation encodée pour générer la traduction mot à mot dans la langue cible (espagnol). Cette architecture est la plus naturelle pour mapper une séquence d'entrée à une séquence de sortie, ce qui est l'essence de la traduction automatique.

### Positional Encoding
- **Purpose: why absolute/relative position info matters.**
  Le mécanisme d'auto-attention des Transformers traite chaque mot d'une séquence indépendamment et n'a pas de concept inhérent de l'ordre des mots. Si la même phrase est présentée dans un ordre de mots différent, l'auto-attention produirait la même sortie. L'encodage positionnel (Positional Encoding) est une technique qui injecte des informations sur la position absolue ou relative de chaque token dans la séquence. Il ajoute un vecteur d'information de position à l'embedding de chaque mot, permettant au modèle de comprendre l'ordre des mots et donc la structure syntaxique et sémantique de la phrase. Sans cela, le modèle perdrait la capacité de distinguer le sens de phrases comme "Le chat mange la souris" et "La souris mange le chat".

- **Example failure without positions.**
  - Sentence: "La pomme a été mangée par l'enfant." (The apple was eaten by the child.)
  - Sentence: "L'enfant a été mangé par la pomme." (The child was eaten by the apple.)
  - **Failure**: Sans encodage positionnel, un Transformer ne serait pas en mesure de distinguer ces deux phrases. Les mots "pomme", "mangée", "enfant" sont les mêmes dans les deux phrases. Si le modèle ne tient pas compte de l'ordre de ces mots, il ne verrait aucune différence dans les rôles d'agent et de patient, et pourrait attribuer un sens absurde à la deuxième phrase ou la confondre avec la première. Le sens (qui fait quoi à qui) dépend entièrement de l'ordre des mots, et l'encodage positionnel est ce qui permet au Transformer de percevoir cet ordre.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- **T=0.2**: Le modèle produira des sorties très déterministes, en choisissant presque toujours les jetons les plus probables. Le texte sera très cohérent, conservateur et prévisible, avec peu de créativité ou de variation. Le risque d'hallucination est faible, mais le texte peut paraître répétitif ou générique.
- **T=1.5**: Le modèle explorera un éventail plus large de jetons possibles, y compris ceux avec une probabilité plus faible. Le texte sera beaucoup plus créatif, varié et potentiellement surprenant. Cependant, cela augmente le risque d'incohérence, d'absurdité ou d'erreurs factuelles (hallucinations).
- **T=1.0**: C'est la température par défaut, où les probabilités sont utilisées telles quelles. Cela offre un équilibre entre la cohérence et la créativité, permettant une certaine diversité dans les sorties tout en maintenant une bonne structure et un sens logique.

### 2) Application Design
- **Bedtime stories (creativity vs coherence)**: Pour des histoires pour enfants, je choisirais une **température modérée à légèrement élevée (par exemple, entre 0.8 et 1.2)**. Une température trop basse rendrait les histoires trop génériques et prévisibles. Une température trop élevée pourrait rendre l'histoire incohérente ou effrayante pour un enfant. L'objectif est d'avoir des rebondissements créatifs et des éléments imaginatifs tout en gardant une trame narrative logique et facile à suivre pour les enfants. Cela permettrait au modèle d'introduire des personnages ou des événements inattendus sans devenir totalement absurde.
- **Financial report summaries (accuracy/reliability)**: Pour des synthèses de rapports financiers, l'exactitude et la fiabilité sont primordiales. Je choisirais une **température très basse (par exemple, entre 0.1 et 0.5)**. L'objectif est de maximiser la fidélité aux faits et de minimiser toute interprétation créative ou hallucination. Une faible température garantit que le modèle sélectionne les mots et phrases les plus probables et factuels, réduisant ainsi le risque d'erreurs qui pourraient avoir de graves conséquences financières. D'autres stratégies de décodage comme le beam search avec un grand nombre de faisceaux seraient également pertinentes ici.

### 3) Temperature & Bias
- **How does temperature influence bias surfacing or dampening?**:
  - **Température basse (T < 1)** tend à **amplifier les biais** présents dans les données d'entraînement. En rendant le modèle plus déterministe, il se conforme davantage aux motifs statistiques les plus fréquents et aux stéréotypes qu'il a appris. Si le corpus d'entraînement contient des associations biaisées (par exemple, "infirmière" est souvent associé au genre féminin), une basse température rendra le modèle plus susceptible de reproduire ce biais.
  - **Température élevée (T > 1)** tend à **atténuer ou masquer les biais** spécifiques en introduisant plus de hasard. En explorant des options moins probables, le modèle peut parfois dévier des stéréotypes les plus évidents. Cependant, cela ne supprime pas le biais sous-jacent dans le modèle, et l'augmentation de la créativité peut aussi générer des biais inattendus ou des contenus totalement non pertinents. Cela dilue la prévisibilité du biais mais ne le corrige pas intrinsèquement.
- **Realistic example**:
  - **Prompt**: "Le PDG est arrivé au bureau. Il/elle a tout de suite..."
  - **Avec une basse température (T=0.2)**, si le modèle a été entraîné sur un corpus où la majorité des PDG mentionnés sont des hommes, il est très probable qu'il continue la phrase avec "...a tout de suite demandé son café" ou "...a commencé la réunion". Cela renforce le stéréotype de genre. Le modèle choisit la suite la plus probable, qui est biaisée.
  - **Avec une haute température (T=1.5)**, le modèle pourrait générer une phrase comme "...a tout de suite fait une danse de la joie" ou "...a salué le poisson rouge". Cela dilue le biais de genre en introduisant de l'aléatoire, mais cela se fait au détriment de la pertinence et de la cohérence. Il pourrait occasionnellement générer une description plus neutre ou féminine, mais ce serait par pur hasard plutôt que par une compréhension débiaisée du rôle.
