1. BERT Tokenization
BERT (Bidirectional Encoder Representations from Transformers) uses a WordPiece tokenizer. It breaks down words into smaller subword units if they are not in its vocabulary. This helps in handling out-of-vocabulary words and reduces the vocabulary size.

Let's tokenize the example text using a pre-trained BERT tokenizer.

```python
# Initialize BERT tokenizer (using a cased model for demonstration)
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-cased')

# Tokenize the example text
bert_tokens = bert_tokenizer.tokenize(example_text)
bert_ids = bert_tokenizer.convert_tokens_to_ids(bert_tokens)

print("Original Text:", example_text)
print("\nBERT Tokens:", bert_tokens)
print("BERT Token IDs:", bert_ids)
print("BERT Vocabulary Size:", bert_tokenizer.vocab_size)
```

2. XLM-RoBERTa Tokenization

XLM-RoBERTa is a multilingual model that uses a SentencePiece tokenizer. It is trained on a much larger and more diverse dataset across many languages, making it suitable for cross-lingual tasks. SentencePiece tokenization is purely data-driven and can handle various languages without requiring pre-segmentation of text.

Let's tokenize the example text using a pre-trained XLM-RoBERTa tokenizer.

```python
# Initialize XLM-RoBERTa tokenizer
xlmroberta_tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')

# Tokenize the example text
xlmroberta_tokens = xlmroberta_tokenizer.tokenize(example_text)
xlmroberta_ids = xlmroberta_tokenizer.convert_tokens_to_ids(xlmroberta_tokens)

print("Original Text:", example_text)
print("\nXLM-RoBERTa Tokens:", xlmroberta_tokens)
print("XLM-RoBERTa Token IDs:", xlmroberta_ids)
print("XLM-RoBERTa Vocabulary Size:", xlmroberta_tokenizer.vocab_size)
```

Key Differences:
BERT (WordPiece): Often splits words like tokenisation into token and ##isation. It's primarily designed for English and has multilingual versions.
XLM-RoBERTa (SentencePiece): Uses a byte-pair encoding (BPE) variant that works directly on raw text. It's designed for multilingual tasks and often includes a leading (underscore followed by space) to indicate the start of a word, even for languages that don't use spaces.

## Tokenisation de texte avec Hugging Face Transformers

Nous allons explorer l'utilisation des `BertTokenizer` et `XLMRobertaTokenizer` pour convertir du texte en entrées tokenisées, comprendre les différents types de jetons et expérimenter la tokenisation de phrases uniques et de paires de phrases.

```python
# 1. Importation des tokeniseurs nécessaires
from transformers import BertTokenizer, XLMRobertaTokenizer

# 2. Chargement des tokeniseurs pré-entraînés
print("Chargement de BertTokenizer...")
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
print("BertTokenizer chargé.")

print("Chargement de XLMRobertaTokenizer...")
xlam_roberta_tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')
print("XLMRobertaTokenizer chargé.")
```

### Expérimentation avec BertTokenizer

```python
# Phrase unique
sentence_1 = "Ceci est un exemple de phrase pour la tokenisation."

print("\n--- BertTokenizer - Tokenisation d'une phrase unique ---")
encoded_bert_single = bert_tokenizer(
    sentence_1,
    add_special_tokens=True,  # Ajouter les tokens spéciaux [CLS] et [SEP]
    max_length=512,           # Tronquer/padder à une longueur maximale de 512
    padding='max_length',     # Padder à la longueur maximale
    truncation=True,          # Activer la troncation
    return_attention_mask=True, # Retourner le masque d'attention
    return_tensors='pt'       # Retourner les tensors PyTorch
)

print("Sentence 1:", sentence_1)
print("Input IDs (Bert):", encoded_bert_single['input_ids'])
print("Attention Mask (Bert):", encoded_bert_single['attention_mask'])
print("Tokens décodés (Bert):", bert_tokenizer.decode(encoded_bert_single['input_ids'][0], skip_special_tokens=False))

# Deux phrases
sentence_2 = "Comment ça va aujourd'hui ?"

print("\n--- BertTokenizer - Tokenisation de deux phrases ---")
encoded_bert_pair = bert_tokenizer(
    sentence_1,
    sentence_2,
    add_special_tokens=True,
    max_length=512,
    padding='max_length',
    truncation=True,
    return_attention_mask=True,
    return_token_type_ids=True, # Important pour la tokenisation de paires de phrases
    return_tensors='pt'
)

print("Sentence 1:", sentence_1)
print("Sentence 2:", sentence_2)
print("Input IDs (Bert):", encoded_bert_pair['input_ids'])
print("Attention Mask (Bert):", encoded_bert_pair['attention_mask'])
print("Token Type IDs (Bert):", encoded_bert_pair['token_type_ids'])
print("Tokens décodés (Bert):", bert_tokenizer.decode(encoded_bert_pair['input_ids'][0], skip_special_tokens=False))

print("\nExplication des Token Type IDs:")
print("0 pour le premier segment (sentence_1 et [CLS], [SEP])")
print("1 pour le second segment (sentence_2 et son [SEP])")
```

### Expérimentation avec XLMRobertaTokenizer

XLMRoberta n'utilise pas `token_type_ids` de la même manière que BERT pour distinguer les paires de phrases. Il utilise généralement un seul type d'ID de token (0) et se fie aux tokens `<s>` et `</s>` pour délimiter les phrases.

```python
# Phrase unique
sentence_1_xlm = "Ceci est un exemple de phrase pour la tokenisation."

print("\n--- XLMRobertaTokenizer - Tokenisation d'une phrase unique ---")
encoded_xlm_single = xlam_roberta_tokenizer(
    sentence_1_xlm,
    add_special_tokens=True,
    max_length=512,
    padding='max_length',
    truncation=True,
    return_attention_mask=True,
    return_tensors='pt'
)

print("Sentence 1:", sentence_1_xlm)
print("Input IDs (XLMRoberta):", encoded_xlm_single['input_ids'])
print("Attention Mask (XLMRoberta):", encoded_xlm_single['attention_mask'])
print("Tokens décodés (XLMRoberta):", xlam_roberta_tokenizer.decode(encoded_xlm_single['input_ids'][0], skip_special_tokens=False))

# Deux phrases
sentence_2_xlm = "Comment ça va aujourd'hui ?"

print("\n--- XLMRobertaTokenizer - Tokenisation de deux phrases ---")
encoded_xlm_pair = xlam_roberta_tokenizer(
    sentence_1_xlm,
    sentence_2_xlm,
    add_special_tokens=True,
    max_length=512,
    padding='max_length',
    truncation=True,
    return_attention_mask=True,
    # XLMRoberta n'utilise généralement pas return_token_type_ids pour la distinction de paires de phrases de la même manière que BERT
    return_tensors='pt'
)

print("Sentence 1:", sentence_1_xlm)
print("Sentence 2:", sentence_2_xlm)
print("Input IDs (XLMRoberta):", encoded_xlm_pair['input_ids'])
print("Attention Mask (XLMRoberta):", encoded_xlm_pair['attention_mask'])
# Notez que 'token_type_ids' peut ne pas être présent ou être rempli de zéros pour XLM-RoBERTa
if 'token_type_ids' in encoded_xlm_pair:
    print("Token Type IDs (XLMRoberta):", encoded_xlm_pair['token_type_ids'])
else:
    print("Token Type IDs (XLMRoberta): Non disponible ou non utilisé pour la distinction de segment de cette manière.")
print("Tokens décodés (XLMRoberta):", xlam_roberta_tokenizer.decode(encoded_xlm_pair['input_ids'][0], skip_special_tokens=False))
```

### Résumé des types de jetons

*   **`input_ids`** : Ce sont les indices numériques (IDs) qui représentent chaque token dans le vocabulaire du modèle. C'est l'entrée principale pour le modèle.
*   **`attention_mask`** : Un tenseur binaire (0 ou 1) qui indique quels tokens doivent être pris en compte par le mécanisme d'attention (1) et lesquels sont des tokens de remplissage (padding) et doivent être ignorés (0).
*   **`token_type_ids`** (spécifique à BERT et modèles similaires pour les tâches de paires de phrases) : Un tenseur binaire (0 ou 1) utilisé pour distinguer les deux segments d'une paire de phrases. Pour BERT, 0 indique le premier segment et 1 le second. XLM-RoBERTa gère la distinction de segment différemment et n'utilise généralement pas ce champ de la même manière.

### Comprendre l'Attention Mask

L'`attention_mask` est crucial pour les modèles de transformateurs. C'est un tenseur de la même forme que les `input_ids`, où:

*   `1` indique un token réel qui doit être traité par le modèle.
*   `0` indique un token de padding (`[PAD]`) qui doit être ignoré par le mécanisme d'attention du modèle. Cela empêche le modèle de gaspiller des ressources de calcul sur des tokens non informatifs et évite que le padding n'affecte les calculs d'attention.

5. Création de plis de validation croisée


Objectif : Mettre en œuvre une validation croisée à k plis pour l'entraînement.

```python
# Part 5: Cross-Validation Fold Creation
from sklearn.model_selection import StratifiedKFold
import numpy as np # Import numpy for np.unique

# Define a seed for reproducibility
seed = 42

# Extract features (text) and labels from the training dataframe
X = train_df["text"].values # Use the 'text' column as features
y = train_df["label"].values # Use the 'label' column for stratification

# Initialize StratifiedKFold with 5 folds
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

# Lists to store train/validation sets for each fold
train_folds = []
val_folds   = []

print("\nCreating stratified cross-validation folds...")
# Create the folds
for fold_idx, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    X_train_fold, X_val_fold = X[train_idx], X[val_idx]
    y_train_fold, y_val_fold = y[train_idx], y[val_idx]

    train_folds.append((X_train_fold, y_train_fold))
    val_folds.append((X_val_fold,   y_val_fold))

    # Verify label distribution per fold
    unique, counts = np.unique(y_train_fold, return_counts=True)
    train_dist = dict(zip(unique, counts / counts.sum()))

    unique, counts = np.unique(y_val_fold, return_counts=True)
    val_dist = dict(zip(unique, counts / counts.sum()))

    print(f"Fold {fold_idx + 1} | "
          f"Train: {len(X_train_fold)} samples {train_dist} | "
          f"Val: {len(X_val_fold)} samples {val_dist}")

print(f"\nTotal folds created → Train sets: {len(train_folds)} | Val sets: {len(val_folds)}")
```

## Vectorisation du texte pour l'entraînement

Nous allons maintenant vectoriser le texte de chaque pli d'entraînement (`train_folds`) en utilisant le `bert_tokenizer`. Cette étape est cruciale pour convertir les données textuelles brutes en un format numérique que le modèle BERT peut traiter.

```python
from transformers import BertTokenizer

# Assurez-vous que le tokenizer est chargé
if 'bert_tokenizer' not in locals():
    print("Chargement de BertTokenizer...")
    bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    print("BertTokenizer chargé.")

tokenized_train_folds = []

print("\nVectorisation du texte pour chaque pli d'entraînement...")

# Itérer sur chaque pli d'entraînement
for fold_idx, (X_train_fold, y_train_fold) in enumerate(train_folds):
    print(f"\n--- Traitement du Pli {fold_idx + 1} ---")

    # Tokenisation du texte du pli d'entraînement
    # Utilisation de padding='max_length' et truncation=True pour uniformiser les longueurs
    # return_tensors='pt' pour retourner des tenseurs PyTorch
    encoded_inputs = bert_tokenizer(
        list(X_train_fold), # Convertir en liste pour le tokenizer
        add_special_tokens=True,
        max_length=128, # Une longueur maximale typique pour BERT
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_token_type_ids=True,
        return_tensors='pt'
    )

    # Stocker les inputs encodés et les labels correspondants
    tokenized_train_folds.append({
        'input_ids': encoded_inputs['input_ids'],
        'attention_mask': encoded_inputs['attention_mask'],
        'token_type_ids': encoded_inputs['token_type_ids'],
        'labels': y_train_fold # Les labels associés à ce pli
    })

    print(f"Pli {fold_idx + 1} - Input IDs (forme):", encoded_inputs['input_ids'].shape)
    print(f"Pli {fold_idx + 1} - Attention Mask (forme):", encoded_inputs['attention_mask'].shape)
    print(f"Pli {fold_idx + 1} - Token Type IDs (forme):", encoded_inputs['token_type_ids'].shape)
    print(f"Pli {fold_idx + 1} - Labels (forme):", y_train_fold.shape)

    # Afficher un exemple de tokenisation pour le premier échantillon du pli
    if fold_idx == 0:
        print("\nExemple pour le premier échantillon du Pli 1:")
        print("Texte original:", X_train_fold[0])
        print("Input IDs:", encoded_inputs['input_ids'][0])
        print("Attention Mask:", encoded_inputs['attention_mask'][0])
        print("Token Type IDs:", encoded_inputs['token_type_ids'][0])
        print("Tokens décodés:", bert_tokenizer.decode(encoded_inputs['input_ids'][0], skip_special_tokens=False))

print("\nVectorisation terminée pour tous les plis d'entraînement.")
print(f"Nombre de plis vectorisés: {len(tokenized_train_folds)}")
```

Les données vectorisées pour chaque pli d'entraînement sont maintenant stockées dans la liste `tokenized_train_folds`. Chaque élément de cette liste est un dictionnaire contenant les `input_ids`, `attention_mask`, `token_type_ids` et les `labels` pour le pli correspondant. Ces données sont prêtes à être utilisées pour l'entraînement d'un modèle de transformateur.
