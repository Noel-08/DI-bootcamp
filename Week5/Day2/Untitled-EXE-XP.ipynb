### 🌟 Exercice 1 : Apprentissage profond vs. Apprentissage automatique traditionnel

### Objectif:
Comprendre les différences entre l'apprentissage automatique traditionnel et l'apprentissage profond.

### 1. Tableau comparatif : Apprentissage automatique traditionnel vs. Apprentissage profond

| Aspect                 | Apprentissage Automatique Traditionnel                       | Apprentissage Profond                                        |
| :--------------------- | :----------------------------------------------------------- | :----------------------------------------------------------- |
| **Ingénierie des fonctionnalités** | Nécessite une ingénierie manuelle et experte des fonctionnalités pour extraire des caractéristiques pertinentes des données brutes. | Réalise automatiquement l'extraction des caractéristiques à travers ses multiples couches, réduisant ainsi le besoin d'ingénierie manuelle. |
| **Informatique**       | Moins exigeant en termes de puissance de calcul ; peut fonctionner efficacement sur des CPU standard. | Très exigeant en puissance de calcul, nécessitant souvent des GPU ou des TPU pour un entraînement efficace. |
| **Évolutivité**         | Les performances tendent à se stabiliser ou à diminuer avec l'augmentation des données après un certain point. | Les performances s'améliorent généralement avec l'augmentation de la quantité de données d'entraînement, montrant une meilleure évolutivité avec de grands ensembles de données. |
| **Découverte de modèles** | Les modèles sont souvent plus interprétables, permettant de comprendre les relations entre les caractéristiques et les sorties. | Les modèles sont des "boîtes noires" complexes, rendant l'interprétabilité des décisions plus difficile en raison des nombreuses couches non linéaires. |
| **Exigences de calcul** | Faibles à modérées pour l'entraînement et l'inférence.       | Très élevées pour l'entraînement, modérées à élevées pour l'inférence selon la complexité du modèle. |


### 2. Identification de problèmes concrets

*   **Problème où l'apprentissage automatique traditionnel est plus adapté :** Prédire le prix d'un logement en fonction de caractéristiques structurées comme la surface, le nombre de pièces, l'emplacement et l'année de construction. Les algorithmes comme la régression linéaire, les arbres de décision ou les forêts aléatoires sont très efficaces avec ce type de données tabulaires et peu nombreuses, et leur interprétabilité est un avantage pour expliquer les facteurs influençant le prix.

*   **Problème où l'apprentissage profond est préférable :** La reconnaissance faciale ou la détection d'objets dans des images. Les données (pixels d'images) sont non structurées et de très grande dimension. Les réseaux de neurones convolutionnels (CNN), une forme d'apprentissage profond, excellent dans l'extraction automatique de caractéristiques hiérarchiques et complexes à partir d'images, surpassant largement les méthodes traditionnelles dans ces tâches.


### 3. Avantage de l'apprentissage profond dans le traitement des données non structurées

L'apprentissage profond excelle dans le traitement des données non structurées (comme les images, le texte, l'audio) car ses architectures multi-couches peuvent apprendre automatiquement des représentations hiérarchiques et des caractéristiques complexes directement à partir des données brutes, sans nécessiter d'ingénierie manuelle fastidieuse. Chaque couche extrait des caractéristiques de plus en plus abstraites, permettant au modèle de comprendre des motifs complexes et des relations profondes. Cette capacité d'apprentissage de bout en bout est cruciale car les données non structurées manquent de schémas clairs et nécessitent une modélisation flexible pour capturer leur richesse inhérente.

### 🌟 Exercice 2 : Réseaux de neurones artificiels (RNA)

### Objectif:
Identifier et comprendre les composantes clés d'un réseau neuronal artificiel.

### 1. Description d'un Réseau Neuronal Artificiel (RNA) simple

Imaginez un RNA simple structuré comme suit :

*   **Couche d'entrée :** Composée de 3 neurones. Ces neurones reçoivent les données brutes (les *caractéristiques*) que le réseau doit traiter. Ils ne réalisent aucune transformation, ils transmettent simplement les valeurs d'entrée.

*   **Couche cachée :** Composée de 4 neurones. C'est ici que les calculs complexes commencent. Chaque neurone de cette couche reçoit des entrées de *tous* les neurones de la couche d'entrée.

*   **Couche de sortie :** Composée de 2 neurones. Ces neurones produisent le résultat final du réseau, comme une prédiction ou une classification.

Graphiquement, cela pourrait être représenté comme trois colonnes de cercles (les neurones), avec des lignes (les connexions) allant de chaque neurone d'une couche à chaque neurone de la couche suivante.

### 2. Composantes du réseau

Voici les composantes clés de ce RNA :

*   **Neurones (ou Nœuds) :** Les unités de traitement fondamentales du réseau, représentées par les cercles dans chaque couche. Chaque neurone reçoit des entrées, effectue un calcul et produit une sortie.

*   **Poids (Weights) :** Chaque connexion entre deux neurones a un poids associé. Ces poids (des valeurs numériques) déterminent l'importance d'une entrée particulière pour le neurone récepteur. Un poids élevé signifie que l'entrée a une forte influence sur la sortie du neurone.

*   **Biais (Bias) :** Un biais est une valeur ajoutée à la somme pondérée des entrées de chaque neurone avant l'application de la fonction d'activation. Il permet d'ajuster la sortie du neurone indépendamment de ses entrées, offrant ainsi plus de flexibilité au modèle pour s'adapter aux données.

*   **Fonction d’activation :** Appliquée à la sortie de chaque neurone (sauf souvent pour les neurones d'entrée) après la somme pondérée et l'ajout du biais. Elle introduit une non-linéarité dans le réseau, permettant au modèle d'apprendre des relations complexes et non linéaires dans les données. Des exemples courants sont ReLU, Sigmoid ou Tanh.

*   **Couches :** Les regroupements de neurones. On distingue la couche d'entrée, les couches cachées (ici une seule), et la couche de sortie.

### 3. Comment l'information circule dans le réseau

L'information circule dans le réseau de manière unidirectionnelle, de la couche d'entrée vers la couche de sortie. Tout d'abord, les données brutes (les *caractéristiques*) sont alimentées aux 3 neurones de la couche d'entrée. Ensuite, chaque neurone de la couche cachée reçoit une somme pondérée des sorties de tous les neurones de la couche d'entrée, à laquelle s'ajoute un biais. Cette somme est ensuite transformée par une fonction d'activation, produisant la sortie du neurone caché. Ce processus se répète pour les neurones de la couche de sortie, qui reçoivent les sorties pondérées des neurones de la couche cachée, ajoutent leur biais, appliquent une fonction d'activation, et finalement produisent les prédictions ou classifications finales du réseau.

### 🌟 Exercice 3 : Création de l’ensemble de données et visualisation des données

### Objectif :
Générer un ensemble de données bruitées et le visualiser.

### 1. Importation des bibliothèques nécessaires

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

### 2. Création de l'ensemble de données bruitées

```
np.random.seed(0) # Pour la reproductibilité
x = np.arange(-1, 1, 0.1) # Crée 20 points de -1 à 0.9 avec un pas de 0.1
y = -x**2 + np.random.normal(0, 0.05, len(x)) # Ajoute du bruit normalement distribué

print("X data points:", x)
print("Y data points:", y)
```

### 3. Représentation des points générés sur un nuage de points

```
fig = plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Données bruitées')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Visualisation des données bruitées (y = -x^2)')
plt.grid(True)
plt.legend()
plt.show()
```

### 4. Division des données en ensembles d'entraînement et de test

```
x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

print(f"Taille de l'ensemble d'entraînement (x_train): {len(x_train)} points")
print(f"Taille de l'ensemble de test (x_test): {len(x_test)} points")

print("x_train:", x_train)
print("y_train:", y_train)
print("x_test:", x_test)
print("y_test:", y_test)
```

### 🌟 Exercice 4 : Ajustement de modèles polynomiaux de différents degrés

### Objectif :
Ajuster des modèles polynomiaux de degrés variables et observer le surapprentissage.

### 1. Définition de la fonction `polynomial_fit`

```
def polynomial_fit(degree):
    """
    Ajuste un polynôme de degré donné aux données d'entraînement (x_train, y_train).
    Retourne le polynôme ajusté (coefficients).
    """
    # np.polyfit renvoie les coefficients du polynôme [p_n, p_{n-1}, ..., p_1, p_0]
    # où p_n * x^n + ... + p_1 * x + p_0
    return np.polyfit(x_train, y_train, degree)
```

### 2. Définition de la fonction `plot_polyfit`

```
def plot_polyfit(degree):
    """
    Trace les données d'entraînement, les données de test et la courbe polynomiale ajustée
    pour un degré donné.
    """
    # Ajuster le polynôme
    poly_coeffs = polynomial_fit(degree)
    # Créer une fonction polynomiale à partir des coefficients
    poly_func = np.poly1d(poly_coeffs)

    # Générer des points pour tracer la courbe du polynôme
    x_plot = np.linspace(x.min(), x.max(), 100)
    y_plot = poly_func(x_plot)

    plt.figure(figsize=(10, 6))
    plt.scatter(x_train, y_train, label='Données d\'entraînement', color='blue', s=50)
    plt.scatter(x_test, y_test, label='Données de test', color='red', s=50)
    plt.plot(x_plot, y_plot, label=f'Polynôme de degré {degree}', color='green', linewidth=2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Ajustement polynomial de degré {degree} sur les données bruitées')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Calculer l'erreur quadratique moyenne (MSE) sur les ensembles d'entraînement et de test
    y_train_pred = poly_func(x_train)
    y_test_pred = poly_func(x_test)

    mse_train = mean_squared_error(y_train, y_train_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)

    print(f"Degré {degree} - MSE entraînement: {mse_train:.4f}")
    print(f"Degré {degree} - MSE test: {mse_test:.4f}")
```

### 3. Visualisation de l'ajustement pour des polynômes de degrés 1, 7 et 11

```
# Degré 1 (sous-apprentissage)
plot_polyfit(1)
```

```
# Degré 7 (bon ajustement potentiel)
plot_polyfit(7)
```

```
# Degré 11 (surapprentissage potentiel)
plot_polyfit(11)
```

### 4. Observation du comportement du modèle (surapprentissage)

En observant les tracés et les valeurs de MSE :

*   **Degré 1 (Modèle linéaire) :** Le modèle de degré 1 est trop simple pour capturer la courbure des données. Il présente une erreur élevée sur les deux ensembles (entraînement et test), ce qui indique un **sous-apprentissage** (ou *underfitting*). La courbe ne suit pas bien la tendance générale des points.

*   **Degré 7 :** Le modèle de degré 7 semble bien capturer la tendance des données. L'erreur d'entraînement est faible et l'erreur de test est également raisonnablement basse. Il représente un bon équilibre entre la complexité du modèle et sa capacité de généralisation.

*   **Degré 11 :** Le modèle de degré 11 s'ajuste presque parfaitement aux points d'entraînement, ce qui se traduit par une très faible MSE d'entraînement. Cependant, il montre des oscillations importantes entre les points et une MSE de test plus élevée. C'est un signe clair de **surapprentissage** (ou *overfitting*) : le modèle a mémorisé le bruit et les spécificités des données d'entraînement au lieu d'apprendre la relation sous-jacente, ce qui le rend moins performant sur de nouvelles données non vues.

### 🌟 Exercice 5 : Validation croisée pour trouver le degré optimal

### Objectif :
Utiliser la validation croisée pour déterminer le degré polynomial optimal.

### 1. Parcourir les polynômes de degré 1 à 11 et calculer la RMSE

```
results = []

for degree in range(1, 12):
    # Ajuster le polynôme
    poly_coeffs = polynomial_fit(degree)
    poly_func = np.poly1d(poly_coeffs)

    # Calculer les prédictions
    y_train_pred = poly_func(x_train)
    y_test_pred = poly_func(x_test)

    # Calculer la RMSE
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    results.append({'degree': degree, 'rmse_train': rmse_train, 'rmse_test': rmse_test})

# Afficher les résultats
print("Degré | RMSE Entraînement | RMSE Test")
print("------------------------------------")
for res in results:
    print(f"{res['degree']:<5} | {res['rmse_train']:.4f}           | {res['rmse_test']:.4f}")
```

### 2. Tracer les valeurs de RMSE en fonction du degré du polynôme

```
degrees = [res['degree'] for res in results]
rmse_train_values = [res['rmse_train'] for res in results]
rmse_test_values = [res['rmse_test'] for res in results]

plt.figure(figsize=(10, 6))
plt.plot(degrees, rmse_train_values, label='RMSE Entraînement', marker='o')
plt.plot(degrees, rmse_test_values, label='RMSE Test', marker='x')
plt.xlabel('Degré du Polynôme')
plt.ylabel('RMSE (échelle logarithmique)')
plt.title('RMSE d\'Entraînement et de Test vs. Degré du Polynôme')
plt.yscale('log') # Utiliser une échelle logarithmique pour l'axe des ordonnées
plt.xticks(degrees) # Afficher tous les degrés sur l'axe des x
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

### 3. Identifier le degré optimal

Pour identifier le degré optimal, nous recherchons le point où la RMSE de l'ensemble de test est minimale avant de commencer à augmenter considérablement, signe de surapprentissage.

*   **Analyse du graphique :** Nous pouvons observer que la RMSE d'entraînement diminue continuellement à mesure que le degré du polynôme augmente (le modèle s'adapte de mieux en mieux aux données d'entraînement). Cependant, la RMSE de test diminue d'abord, atteint un minimum, puis commence à augmenter.

*   **Degré optimal :** En inspectant les valeurs et le graphique, le degré qui minimise la RMSE sur l'ensemble de test semble être le **Degré 2**. C'est cohérent avec le modèle sous-jacent qui est `y = -x^2`, un polynôme de second degré. Au-delà de ce degré, le modèle commence à "mémoriser" le bruit présent dans les données d'entraînement, ce qui dégrade ses performances sur l'ensemble de test (surapprentissage).
