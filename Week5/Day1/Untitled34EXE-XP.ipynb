🌟 Exercice 1 : Petit quiz

1. ML traditionnel vs apprentissage profond
Le ML traditionnel repose sur une ingénierie manuelle des caractéristiques : un expert humain sélectionne et transforme les variables pertinentes avant de les passer à un algorithme (SVM, régression, arbre de décision…). L'apprentissage profond apprend automatiquement ces représentations à travers des couches successives de neurones — aucune extraction manuelle n'est nécessaire.
2. Comment les RNA imitent le cerveau
Un réseau de neurones artificiels s'inspire du neurone biologique : chaque nœud reçoit des signaux (entrées pondérées), les somme, puis applique une fonction d'activation pour décider si le signal se propage. Les connexions synaptiques correspondent aux poids, ajustés lors de l'apprentissage comme les synapses se renforcent par répétition.
3. Pourquoi le deep learning performe mieux sur de grands datasets
Les algorithmes classiques atteignent rapidement un plateau : au-delà d'une certaine quantité de données, leurs performances stagnent. Le deep learning continue de s'améliorer avec le volume, car ses nombreuses couches capturent des abstractions de plus en plus riches. Il exploite réellement la richesse des grands corpus.
4. Défis et solutions

5. Ingénierie des caractéristiques et deep learning
L'ingénierie des caractéristiques consiste à créer manuellement des variables explicatives pertinentes à partir des données brutes (ex. : extraire le jour de la semaine d'une date). En deep learning, les couches cachées apprennent automatiquement ces représentations hiérarchiques — les premières couches détectent des patterns simples (bords, fréquences), les suivantes des abstractions complexes (visages, mots).
6. Rôle des couches cachées
Elles transforment progressivement les données d'entrée en représentations de plus en plus abstraites. Chaque couche extrait des caractéristiques non linéaires que la suivante combine. Sans couches cachées, le réseau ne serait qu'une régression linéaire.
7. Fonction d'activation
Elle introduit la non-linéarité dans le réseau. Sans elle, l'empilement de couches resterait équivalent à une seule transformation linéaire. Elle détermine si un neurone "s'active" et avec quelle intensité — ReLU, sigmoid et softmax en sont les exemples clés.



## 🌟 Exercice 2 : Construction d'un système de décision perceptron simple

Vous allez créer manuellement un système de décision simple basé sur un perceptron pour déterminer si vous devez sortir en fonction de deux entrées :

- Température (°F)
- Pluvieux (Oui = 1, Non = 0)

Vous attribuerez des pondérations, calculerez la somme pondérée, appliquerez une fonction d'activation et déterminerez la décision finale.

**Paramètres :**
- Poids de la température = 0,6
- Poids de la pluie = 0,4
- Biais = 2

**Formule de la somme pondérée :**

`Somme pondérée = (Température × 0,6) + (Pluie × 0,4) + Biais`

**Fonction d'activation par étape :**

- Si la somme pondérée est supérieure à 20, afficher 1 (Oui, sortir).
- Si la somme pondérée est inférieure ou égale à 20, afficher 0 (Non, restez à l'intérieur)


```python
# Définition des poids et du biais
weight_temperature = 0.6
weight_rain = 0.4
bias = 2

def decide_to_go_out(temperature, rain):
    # Calcul de la somme pondérée
    weighted_sum = (temperature * weight_temperature) + (rain * weight_rain) + bias
    print(f"  Somme pondérée: {weighted_sum:.2f}")

    # Application de la fonction d'activation
    if weighted_sum > 20:
        return 1 # Oui, sortir
    else:
        return 0 # Non, restez à l'intérieur
```

### 1. Calculez la sortie pour les conditions suivantes :

**Cas 1 : Température = 70°F, Pluie = 0 (Non)**

```python
temperature_case1 = 70
rain_case1 = 0

print(f"Cas 1: Température = {temperature_case1}°F, Pluie = {rain_case1}")
decision_case1 = decide_to_go_out(temperature_case1, rain_case1)
print(f"  Décision: {'Oui, sortir' if decision_case1 == 1 else 'Non, rester à l\'intérieur'} (Code: {decision_case1})\n")
```

**Cas 2 : Température = 50°F, Pluie = 1 (Oui)**

```python
temperature_case2 = 50
rain_case2 = 1

print(f"Cas 2: Température = {temperature_case2}°F, Pluie = {rain_case2}")
decision_case2 = decide_to_go_out(temperature_case2, rain_case2)
print(f"  Décision: {'Oui, sortir' if decision_case2 == 1 else 'Non, rester à l\'intérieur'} (Code: {decision_case2})\n")
```

### 2. Interprétez les résultats : le perceptron a-t-il suggéré de sortir dans les deux cas ? Pourquoi ou pourquoi pas ?

**Interprétation :**

- **Cas 1 (Température = 70°F, Pluie = 0) :**
  La somme pondérée est calculée comme `(70 * 0.6) + (0 * 0.4) + 2 = 42 + 0 + 2 = 44`. Comme `44` est supérieur à `20`, le perceptron suggère de **sortir** (Décision : 1). Une température élevée et l'absence de pluie incitent à sortir.

- **Cas 2 (Température = 50°F, Pluie = 1) :**
  La somme pondérée est calculée comme `(50 * 0.6) + (1 * 0.4) + 2 = 30 + 0.4 + 2 = 32.4`. Comme `32.4` est supérieur à `20`, le perceptron suggère également de **sortir** (Décision : 1).

Dans les deux cas, le perceptron a suggéré de sortir. Cela est dû au fait que la somme pondérée, même avec une température plus basse et de la pluie dans le Cas 2, reste supérieure au seuil de 20. Les poids attribués (0.6 pour la température) et le biais (2) sont suffisamment élevés pour que la sortie pondérée dépasse le seuil dans ces scénarios. Le poids de la température a un impact plus important que celui de la pluie, et le biais ajoute une valeur de base qui rend la somme pondérée plus facile à dépasser le seuil.

## 🌟 Exercice 3 : Création d'un réseau neuronal simple avec TensorFlow/Keras

Concevez un réseau de neurones simple à l'aide de TensorFlow/Keras pour classifier les chiffres manuscrits du jeu de données MNIST. Le réseau doit comporter :

- Une seule couche d'entrée.
- Une couche cachée avec 128 neurones et une activation ReLU.
- Une couche de sortie avec 10 neurones (pour 10 classes) et une activation softmax.

**Mesures :**

1. Importez TensorFlow et les modules nécessaires (`layers`, `models` et `utilitaires`) depuis `tensorflow.keras`.
2. Chargez le jeu de données MNIST.
3. Normalisez les données en ramenant les valeurs des pixels des images à l'intervalle [0, 1] par 255.
4. Encodez les étiquettes en one-hot en convertissant les étiquettes numériques (0 à 9) en vecteurs one-hot.
5. Pour construire le modèle de réseau de neurones, créez un modèle séquentiel avec l'architecture suivante :
   - Une couche `Flatten` pour convertir chaque image 28x28 en un vecteur 1D.
   - Une couche cachée `Dense` avec 128 neurones et une activation `ReLU`.
   - Une couche de sortie `Dense` avec 10 neurones et une activation `softmax` (pour les 10 classes).
6. Compilez le modèle avec les paramètres suivants :
   - Optimiseur : `'adam'`
   - Fonction de perte : `'categorical_crossentropy'`
   - Métriques : `['accuracy']`
7. Entraînez le modèle et évaluez-le sur l'ensemble de données de test.

**Jeu de données :** Le jeu de données MNIST est inclus dans TensorFlow/Keras et peut être chargé à l’aide de :

`from tensorflow.keras.datasets import mnist`

### 1. Importez TensorFlow et les modules nécessaires

```python
import tensorflow as tf
from tensorflow.keras import layers, models, utils

print(f"TensorFlow version: {tf.__version__}")
```

### 2. Chargez le jeu de données MNIST

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

### 3. Normalisez les données

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

print(f"x_train normalized shape: {x_train.shape}")
print(f"x_test normalized shape: {x_test.shape}")
```

### 4. Encodez les étiquettes en one-hot

```python
num_classes = 10
y_train_one_hot = utils.to_categorical(y_train, num_classes)
y_test_one_hot = utils.to_categorical(y_test, num_classes)

print(f"y_train one-hot shape: {y_train_one_hot.shape}")
print(f"y_test one-hot shape: {y_test_one_hot.shape}")
```

### 5. Pour construire le modèle de réseau de neurones

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)), # Couche d'entrée implicite
    layers.Dense(128, activation='relu'), # Couche cachée avec 128 neurones et ReLU
    layers.Dense(num_classes, activation='softmax') # Couche de sortie avec 10 neurones et softmax
])

model.summary()
```

### 6. Compilez le modèle

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

### 7. Entraînez le modèle et évaluez-le

```python
history = model.fit(
    x_train,
    y_train_one_hot,
    epochs=5,
    batch_size=32,
    validation_split=0.2 # Utiliser 20% des données d'entraînement pour la validation
)

print("\nÉvaluation sur l'ensemble de test:")
loss, accuracy = model.evaluate(x_test, y_test_one_hot, verbose=0)
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

## 🌟 Exercice 4 : Calcul de propagation directe

Dans cet exercice, vous calculerez manuellement la propagation avant d'un réseau neuronal simple qui prédit les prix des maisons en fonction de :

- Superficie en pieds carrés (x₁)
- Nombre de chambres (x₂)

Nous calculerons le résultat en utilisant les valeurs suivantes :

**Input Values:**
- `x₁ = 2000` (Square Footage)
- `x₂ = 3` (Number of Bedrooms)

**Initial Weights:**
- `w₁ = 0.5` (Weight for Square Footage)
- `w₂ = 0.7` (Weight for Bedrooms)

**Bias:** `b = 50,000`

**Activation Function:** ReLU (Rectified Linear Unit)

**Étapes :**
1. Calculez la valeur de sortie « z » avant l’activation.
2. Appliquez la fonction ReLU pour calculer la prédiction finale.
3. Interprétez le résultat : quel est le prix de la maison prévu ?

### 1. Calculez la valeur de sortie « z » avant l’activation.

```python
# Input Values
x1 = 2000  # Square Footage
x2 = 3     # Number of Bedrooms

# Initial Weights
w1 = 0.5   # Weight for Square Footage
w2 = 0.7   # Weight for Bedrooms

# Bias
bias = 50000

# Calculate the weighted sum (z)
z = (x1 * w1) + (x2 * w2) + bias

print(f"Valeur de z (somme pondérée avant activation): {z}")
```

### 2. Appliquez la fonction ReLU pour calculer la prédiction finale.

```python
# Activation Function: ReLU (Rectified Linear Unit)
def relu(x):
    return max(0, x)

# Apply the ReLU function to z
predicted_price = relu(z)

print(f"Prix prédit de la maison (après activation ReLU): {predicted_price}")
```

### 3. Interprétez le résultat : quel est le prix de la maison prévu ?

**Interprétation :**

- **Calcul de la somme pondérée (z) :**
  `z = (2000 * 0.5) + (3 * 0.7) + 50000`
  `z = 1000 + 2.1 + 50000`
  `z = 51002.1`

- **Application de la fonction d'activation ReLU :**
  La fonction ReLU retourne `max(0, z)`. Puisque `z = 51002.1` est supérieur à 0, la fonction ReLU retourne `51002.1`.

- **Prix de la maison prévu :**
  Le prix prévu de la maison est **51002.1**.

## 🌟 Exercice 5 (facultatif) : Implémentation de la propagation avant et arrière en Python

Vous allez coder un réseau neuronal simple qui effectue la propagation avant et la rétropropagation pour un problème de régression (prédiction des notes d'examen en fonction des heures d'étude).

```python
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```

### Observez comment les poids et les biais se mettent à jour.

Après l'exécution du code, vous pouvez voir la `Initial Prediction`, la `Loss`, et les `Updated Weights` et `Updated Bias`.

- **Initial Prediction:** C'est la prédiction du modèle avant tout ajustement, basée sur les poids et le biais initiaux.
- **Loss:** C'est la mesure de l'erreur du modèle. Plus la perte est faible, meilleure est la prédiction.
- **Updated Weights & Bias:** Ces valeurs montrent comment les poids et le biais ont été ajustés après une étape de rétropropagation. L'objectif est de réduire la perte en ajustant ces paramètres.

### Expliquez pourquoi la mise à jour des poids à l'aide de la descente de gradient réduit l'erreur.

La **descente de gradient** est un algorithme d'optimisation utilisé pour minimiser la fonction de perte d'un modèle. Voici comment elle réduit l'erreur :

1.  **Calcul du gradient :** Le gradient de la fonction de perte par rapport à chaque poids et biais indique la direction de la plus forte augmentation de l'erreur. Si le gradient est positif, augmenter le poids augmente l'erreur ; s'il est négatif, augmenter le poids diminue l'erreur.
2.  **Mise à jour dans la direction opposée :** Pour minimiser l'erreur, nous voulons nous déplacer dans la direction opposée au gradient. C'est pourquoi nous soustrayons un multiple du gradient des poids et des biais (`w_new = w - learning_rate * grad_w`).
3.  **Taux d'apprentissage :** Le `learning_rate` (taux d'apprentissage) détermine la taille des pas effectués dans la direction opposée au gradient. Un petit taux d'apprentissage conduit à des mises à jour lentes mais potentiellement plus stables, tandis qu'un grand taux d'apprentissage peut accélérer la convergence mais risque de dépasser le minimum (surapprentissage).

En répétant ce processus sur de nombreuses itérations (époques), le modèle ajuste progressivement ses poids et ses biais pour que la prédiction (`y_pred`) se rapproche de la vraie valeur (`y_true`), ce qui réduit la fonction de perte.

0. Importer les bibliothèques requises : Importer TensorFlow, NumPy, Matplotlib et les autres modules Keras nécessaires.

4. Construire le modèle : Créer un modèle avec une couche Flatten pour convertir chaque image en un vecteur, une couche dense avec 128 neurones et une activation ReLU, et une couche dense de sortie avec 10 neurones et une activation softmax.

### 0. Importez les bibliothèques requises (TensorFlow, NumPy, Matplotlib sont déjà importés dans les exercices précédents, mais nous ajoutons explicitement Matplotlib ici pour la visualisation).

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, utils

print("Libraries imported.")
```

### 1-6. (Re)Chargement, Prétraitement, Construction, Compilation et Entraînement du modèle

Nous allons réutiliser le code des étapes 1 à 6 de l'exercice 3 pour s'assurer que le modèle est entraîné avant de faire des prédictions.

```python
# 1. Charger le jeu de données MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normaliser les données
x_train_normalized = x_train.astype('float32') / 255.0
x_test_normalized = x_test.astype('float32') / 255.0

# 3. Encoder les étiquettes en one-hot
num_classes = 10
y_train_one_hot = utils.to_categorical(y_train, num_classes)
y_test_one_hot = utils.to_categorical(y_test, num_classes)

# 4. Construire le modèle
model_viz = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

# 5. Compiler le modèle
model_viz.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 6. Entraîner le modèle
print("Entraînement du modèle...")
history_viz = model_viz.fit(
    x_train_normalized,
    y_train_one_hot,
    epochs=3, # Réduire le nombre d'époques pour un entraînement plus rapide pour la visualisation
    batch_size=32,
    validation_split=0.2,
    verbose=0 # Supprimer la sortie détaillée de l'entraînement
)

print("Modèle entraîné.")
```

### 7. Effectuer des prédictions

```python
predictions = model_viz.predict(x_test_normalized)
predicted_labels = np.argmax(predictions, axis=1)

print("Prédictions effectuées.")
```

### 8. Visualiser quelques prédictions

```python
plt.figure(figsize=(10, 10))
for i in range(25): # Afficher les 25 premières images de test
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i], cmap=plt.cm.binary)

    true_label = y_test[i]
    pred_label = predicted_labels[i]

    color = 'green' if pred_label == true_label else 'red'
    plt.xlabel(f"Préd: {pred_label} (Vrai: {true_label})", color=color)
plt.suptitle('Prédictions du modèle sur le jeu de données MNIST (Vert: Correct, Rouge: Incorrect)', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Ajuster pour éviter le chevauchement du titre
plt.show()
```
