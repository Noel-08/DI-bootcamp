Exercice 1 : Opérations matriciales

```
import numpy as np

# Création d'une matrice 3x3 aléatoire
A = np.array([[4, 7, 2],
              [3, 6, 1],
              [2, 5, 1]])

# 1. Calcul du déterminant
det_A = np.linalg.det(A)

print(f"Matrice A :\n{A}\n")
print(f"Déterminant de A : {det_A:.4f}")

# 2. Calcul de l'inverse (si le déterminant n'est pas nul)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print(f"\nInverse de A :\n{inv_A}")
else:
    print("\nLa matrice n'est pas inversible (déterminant nul).")
```

Exercice 4 : Manipulation de données avec NumPy et Pandas

```
import pandas as pd
import numpy as np

# Création d'un DataFrame 10x3 avec des nombres aléatoires
np.random.seed(42)
df = pd.DataFrame(np.random.randint(10, 100, size=(10, 3)), columns=['A', 'B', 'C'])
print("DataFrame Original :")
print(df, "\n")

# 1. Sélection conditionnelle : Lignes où la colonne A est supérieure à 50
conditional_selection = df[df['A'] > 50]
print("Sélection conditionnelle (A > 50) :")
print(conditional_selection, "\n")

# 2. Agrégations (Somme et Moyenne par colonne)
print("Somme par colonne :")
print(df.sum())
print("\nMoyenne par colonne :")
print(df.mean())
```

```
import numpy as np
import matplotlib.pyplot as plt

# Création d'une image 5x5 en niveaux de gris (valeurs de 0 à 255)
# Diagonale blanche (255) sur fond gris foncé (50)
image_array = np.array([
    [255, 50,  50,  50,  50],
    [50,  255, 50,  50,  50],
    [50,  50,  255, 50,  50],
    [50,  50,  50,  255, 50],
    [50,  50,  50,  50,  255]
])

print("Représentation matricielle NumPy de l'image :")
print(image_array)

# Visualisation graphique de la matrice
plt.imshow(image_array, cmap='gray', vmin=0, vmax=255)
plt.title("Image 5x5 affichée")
plt.colorbar()
plt.show()
```

Exercice 6 : Test d’hypothèses de baseFormulation de l'hypothèse :Hypothèse Nulle (\(H_{0}\)) : Le programme de formation n'a aucun effet (la différence moyenne de productivité avant/après est égale à 0).Hypothèse Alternative (\(H_{a}\)) : Le programme augmente la productivité (la différence moyenne avant/après est supérieure à 0).

```
import numpy as np
from scipy import stats  # Importation recommandée pour les tests statistiques

np.random.seed(42)
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Calcul de la différence
differences = productivity_after - productivity_before
mean_diff = np.mean(differences)

print(f"Amélioration moyenne constatée : {mean_diff:.2f} points.\n")

# Test T apparié (avant vs après sur les mêmes individus)
t_stat, p_val = stats.ttest_rel(productivity_after, productivity_before, alternative='greater')

print(f"Statistique T : {t_stat:.4f}")
print(f"P-value       : {p_val:.6f}")

alpha = 0.05
if p_val < alpha:
    print("\nConclusion : On rejette H0. Le programme de formation est statistiquement efficace.")
else:
    print("\nConclusion : Échec du rejet de H0. On ne peut pas prouver l'efficacité du programme.")
```

Exercice 10 : Visualisation de base

```
import numpy as np
import matplotlib.pyplot as plt

# Génération des données NumPy (Évolution cumulative de valeurs aléatoires)
np.random.seed(10)
x = np.arange(1, 51)
y = np.random.randn(50).cumsum()

# Création du graphique linéaire avec Matplotlib
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Tendance aléatoire", color="blue", linewidth=2, marker='o', markersize=4)

# Personnalisation
plt.title("Graphique linéaire de nombres aléatoires cumulés", fontsize=14)
plt.xlabel("Index / Temps", fontsize=12)
plt.ylabel("Valeurs", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Affichage dans Colab
plt.show()
```
