1. Énoncé du problème
Problème métier :
Les institutions financières subissent des pertes importantes dues aux défauts de paiement sur les prêts accordés. Actuellement, les décisions d’octroi de crédit reposent largement sur des règles manuelles et des scores de crédit traditionnels (comme le FICO), qui ne capturent pas toujours les patterns complexes et non linéaires présents dans les données modernes. Cela entraîne un taux de faux positifs (refus de bons candidats) et de faux négatifs (octroi à des candidats à haut risque).
Objectif du projet :
Développer un modèle de machine learning capable de prédire avec une grande précision la probabilité qu’un emprunteur entre en défaut de paiement dans les 12 à 24 mois suivant l’octroi du prêt. Le modèle permettra aux banques et fintechs de :
Améliorer la qualité du portefeuille de prêts,
Réduire les pertes liées aux défauts,
Offrir des conditions plus compétitives aux bons profils (taux d’intérêt ou montants plus favorables),
Respecter les exigences réglementaires en matière de risque (Bâle III/IV, RGPD, etc.).
Question analytique principale :
Étant donné les caractéristiques d’un demandeur de prêt au moment de la demande (données démographiques, financières, comportementales et historiques), quelle est la probabilité qu’il soit en défaut dans les X prochains mois ?
Type de problème : Classification binaire (défaut = 1 / non-défaut = 0) avec possibilité d’approche probabiliste (score de risque entre 0 et 1) pour permettre une calibration et une interprétabilité métier.

Instructions

À partir de cet ensemble de données : Prédiction de prêts , identifiez les caractéristiques qui pourraient être les plus pertinentes pour prédire les défauts de paiement de prêts.
Justifiez votre choix de fonctionnalités.

Fonctionnalités principales recommandées :
Credit_History (Historique de crédit)
Pourquoi la plus importante ? C’est généralement la variable la plus prédictive. Un historique de crédit positif (remboursements passés sans incident) est fortement corrélé à une faible probabilité de défaut. À l’inverse, des retards ou défauts antérieurs augmentent fortement le risque. De nombreuses études et modèles (Random Forest, XGBoost, etc.) la placent en tête des features par importance.
ApplicantIncome et CoapplicantIncome (Revenus du demandeur et co-demandeur)
Justification : Plus les revenus sont élevés et stables, plus la capacité de remboursement est forte. Un faible revenu augmente le risque de défaut, surtout si le ratio prêt/revenu est élevé. La combinaison des deux revenus donne une vue plus réaliste de la situation financière du foyer.
LoanAmount (Montant du prêt) et Loan_Amount_Term (Durée du prêt)

Justification : Un montant élevé ou une durée longue augmente la charge de remboursement et donc le risque de défaut. Il est utile de créer des features dérivées comme le ratio LoanAmount / ApplicantIncome (souvent très prédictif).
Property_Area (Zone de la propriété : Urban / Semiurban / Rural)
Justification : Les zones urbaines ou semi-urbaines sont souvent associées à de meilleures opportunités économiques et donc à un risque moindre, tandis que les zones rurales peuvent présenter plus d’instabilité.
Education et Self_Employed
Justification : Un niveau d’études élevé (Graduate) est généralement lié à de meilleurs emplois et revenus stables. Les travailleurs indépendants (Self_Employed) peuvent avoir des revenus plus variables et donc un risque légèrement plus élevé.
Fonctionnalités secondaires / contextuelles (à inclure selon disponibilité) :
Married et Dependents : Indiquent la stabilité familiale et la charge financière (plus d’enfants = potentiellement plus de dépenses).
Gender : Parfois corrélé (dans certains datasets/contextes culturels), mais souvent moins prédictif et sujet à biais éthique → à utiliser avec prudence.
Variables dérivées utiles :
Total_Income = ApplicantIncome + CoapplicantIncome
Loan_to_Income_Ratio
EMI (mensualité estimée)
Fonctionnalités à éviter ou à traiter avec soin :
Loan_ID : Identifiant unique, inutile pour la prédiction (overfitting si conservé).
Variables trop corrélées (multicolinéarité) : par exemple, si vous avez beaucoup de variables de revenu, en garder une ou deux synthétiques.
Justification globale du choix :
Critères de sélection : Pertinence métier (capacité de remboursement, historique de crédit, stabilité), corrélation avec la cible (Loan_Status / Default), et importance dans les modèles ML (feature importance via Random Forest/XGBoost).

Modèles à privilégier :
XGBoost / LightGBM (excellents en performance, gèrent bien les données mixtes et les relations non linéaires).
Random Forest (interprétable via feature importance, robuste).
Logistic Regression (baseline interprétable pour expliquer le business).

Exercice 3 : Entraînement, évaluation et optimisation du modèle

Instructions

Quel(s) modèle(s) choisiriez-vous pour une prédiction de prêt ?
Décrivez les étapes permettant d'évaluer les performances du modèle, en mentionnant les indicateurs spécifiques pertinents pour cette évaluation.

1. Quel(s) modèle(s) choisiriez-vous pour une prévision de prêt ?
La prévision de prêt est typiquement un problème de classification binaire (ex. : risque de défaut = 1 / pas de défaut = 0) ou de régression de probabilité (score de crédit / probabilité de défaut). Le choix des modèles dépend de la taille du jeu de données, du besoin d’interprétabilité (réglementation bancaire) et des performances.
Modèles recommandés (par ordre de priorité) :
Gradient Boosting (les plus performants actuellement)
XGBoost, LightGBM ou CatBoost
→ Meilleur rapport performance/vitesse sur données tabulaires (très utilisé en crédit scoring).
Avantages : gèrent bien les données manquantes, les variables catégorielles, les interactions non linéaires, et donnent d’excellents scores AUC.
Random Forest
Très robuste, moins sensible au surapprentissage que les arbres simples, bonne baseline.
Logistic Regression (avec régularisation L1/L2)
Modèle de base obligatoire dans le secteur bancaire pour son interprétabilité (coefficients = importance des variables) et sa simplicité réglementaire (modèle explicable).
Autres modèles selon le contexte
HistGradientBoosting (scikit-learn) : très rapide.
Neural Networks (MLP, TabNet, ou modèles profonds) : si très gros volume de données (> 100k–1M lignes) et features riches.
SVM ou Naive Bayes : rarement les meilleurs, mais utiles en ensemble.
Stratégie recommandée :
Commencer par Logistic Regression + Random Forest comme baselines.
Passer rapidement à LightGBM / XGBoost / CatBoost.
Faire un stacking ou blending des meilleurs modèles pour gagner encore en performance.
2. Étapes pour évaluer le rendement du modèle
Voici un processus rigoureux et standard en scoring de crédit :
Étape 1 : Préparation des données
Division du dataset : Train / Validation / Test (ex. 60% / 20% / 20%) ou Time-based split si données temporelles (très important en crédit pour éviter le data leakage).
Stratification sur la variable cible (pour conserver la proportion de défauts).
Étape 2 : Entraînement et validation croisée
Utiliser Stratified K-Fold Cross-Validation (5 ou 10 folds) sur le train set.
Optimisation des hyperparamètres avec Bayesian Optimization (Optuna, Hyperopt) ou Grid/Random Search.
Métrique principale d’optimisation : ROC-AUC ou PR-AUC (selon déséquilibre des classes).
Étape 3 : Évaluation sur le jeu de test (hold-out)
Mesures précises pertinentes pour un modèle de prêt :
A. Discriminante (capacité à séparer bons et mauvais clients)
ROC-AUC (Area Under the ROC Curve) → la plus importante (cible > 0.75–0.85 selon métier).
PR-AUC (Precision-Recall AUC) → plus pertinente si forte classe déséquilibrée (défauts rares).
Gini Coefficient = 2 × AUC - 1 (très utilisé dans l’industrie du crédit).
KS Statistic (Kolmogorov-Smirnov) → mesure la séparation maximale entre les distributions de scores des bons et mauvais clients (cible souvent > 0.3–0.4).
B. Mesures à seuil (business-oriented)
Matrice de confusion (avec différents seuils)
Precision (quand on accorde un prêt, quel est le risque réel ?)
Recall (taux de détection des vrais défauts)
F1-score ou F-beta score (si on veut privilégier la précision ou le rappel)
Accuracy (à utiliser avec prudence, trompeuse en cas de déséquilibre)
C. Mesures de calibration (fiabilité des probabilités)
Brier Score
Log Loss (Binary Cross-Entropy)
Courbe de calibration + Expected Calibration Error (ECE)
D. Stabilité et robustesse
PSI (Population Stability Index) : pour vérifier si la distribution des scores change entre train et test / population actuelle.
Feature importance + SHAP values (pour interprétabilité métier et détection de biais).
Tests sur sous-populations (par âge, région, type de prêt, etc.) pour détecter les biais.
Étape 4 : Comparaison et choix final
Comparer plusieurs modèles sur les métriques ci-dessus.
Choisir le modèle selon un compromis Performance / Interprétabilité / Vitesse / Coût de maintenance.
Valider économiquement : simuler l’impact business (ex. : gain/perte selon le seuil de scoring choisi).
Outils pratiques :
scikit-learn, xgboost, lightgbm, catboost
Optuna ou Hyperopt pour l’optimisation
SHAP, LIME ou interpretML pour l’explicabilité
scikit-plot ou yellowbrick pour les visualisations
Ce processus permet d’obtenir un modèle à la fois performant et conforme aux exigences réglementaires (ex. : explicabilité exigée par Bâle III / RGPD).


 Exercice 4 : Concevoir des solutions d'apprentissage automatique pour des problèmes spécifiques

Instructions

Pour chacun de ces scénarios, déterminez quel type d'apprentissage automatique serait le plus approprié. Expliquez votre choix.

Prévision des cours boursiers : prédire les cours futurs
Organiser une bibliothèque : regrouper les livres par genres ou catégories en fonction de leurs similitudes.
Programmez un robot pour qu'il navigue et trouve le chemin le plus court dans un labyrinthe.

1. Prédire les cours des actions : prévoir les prix futurs
Type recommandé : Apprentissage supervisé (Supervised Learning), plus précisément une tâche de régression (Regression).
Explication :
On dispose généralement de données historiques étiquetées (prix passés, volumes, indicateurs techniques, données macroéconomiques, etc.) pour entraîner le modèle.
L’objectif est de prédire une valeur continue (le prix futur), ce qui correspond à un problème de régression.
Modèles couramment utilisés :
Régression linéaire/multiple, Random Forest, Gradient Boosting (XGBoost, LightGBM).
Modèles séquentiels pour les séries temporelles : LSTM, GRU, Transformer (Time Series).
L’apprentissage se fait en minimisant l’erreur entre les prédictions et les prix réels observés (ex. : Mean Squared Error).
Note : Bien que très utilisé, ce type de prédiction reste difficile en raison de l’aspect chaotique et influencé par de nombreux facteurs externes (nouvelles, sentiment du marché). On combine souvent du ML avec de l’analyse fondamentale et technique.
2. Organiser une bibliothèque de livres : regrouper les livres en genres ou catégories en fonction des similitudes
Type recommandé : Apprentissage non supervisé (Unsupervised Learning), plus précisément le Clustering.
Explication :
Il n’y a pas d’étiquettes préalables (genres déjà connus) ; on veut découvrir des groupes naturels à partir des caractéristiques des livres.
On représente chaque livre sous forme vectorielle (ex. : embeddings de texte via TF-IDF, Word2Vec, BERT, ou caractéristiques comme auteur, résumé, mots-clés).
Algorithmes adaptés :
K-Means (simple et efficace).
DBSCAN (gère les clusters de formes variées et le bruit).
Hierarchical Clustering (pour une structure hiérarchique de genres).
Réduction de dimensionnalité préalable (PCA, t-SNE, UMAP) pour visualiser et améliorer les regroupements.
C’est typiquement un problème de clustering ou de topic modeling (LDA, NMF) si on travaille sur le contenu textuel.
3. Programmer un robot pour naviguer et trouver le chemin le plus court dans un labyrinthe
Type recommandé : Apprentissage par renforcement (Reinforcement Learning - RL).
Explication :
Le robot (agent) interagit avec l’environnement (le labyrinthe) sans connaître à l’avance le chemin optimal.
À chaque étape, il reçoit une récompense (reward) positive lorsqu’il avance vers la sortie et une pénalité (négative) s’il heurte un mur ou prend trop de temps.
L’objectif est d’apprendre une politique (policy) qui maximise la récompense cumulative, c’est-à-dire le chemin le plus court et le plus sûr.
Algorithmes adaptés :
Q-Learning ou Deep Q-Network (DQN) pour des labyrinthes discrets.
Policy Gradient / Actor-Critic (A2C, PPO) pour des environnements plus complexes.
A* ou algorithmes de recherche de chemin classiques peuvent être utilisés en complément ou comme baseline, mais le RL permet au robot d’apprendre à s’adapter à des labyrinthes changeants ou partiellement inconnus.
Avantage du RL : Le robot peut généraliser et s’améliorer avec l’expérience (exploration vs exploitation), ce qui est idéal pour des environnements dynamiques.





Exercice 5 : Concevoir une stratégie d'évaluation pour différents modèles ML

Instructions

Sélectionnez trois types de modèles d'apprentissage automatique : un à partir de l'apprentissage supervisé (par exemple, un modèle de classification), un à partir de l'apprentissage non supervisé (par exemple, un modèle de clustering) et un à partir de l'apprentissage par renforcement. Pour le modèle supervisé, esquissez une stratégie pour évaluer ses performances, y compris le choix de métriques (comme la précision, la précision, le rappel, le score F1) et de méthodes (comme la validation croisée, les courbes ROC).
Pour le modèle non supervisé, décrivez comment vous évalueriez l'efficacité du modèle, en tenant compte de techniques telles que le score de silhouette, la méthode du coude ou les mesures de validation de cluster.
Pour le modèle d'apprentissage par renforcement, discutez de la manière dont vous mesureriez son succès, en considérant des aspects tels que la récompense cumulative, la convergence et l'équilibre entre exploration et exploitation.
Relever les défis et les limites de l'évaluation des modèles dans chaque catégorie.



 1. Modèle d'apprentissage supervisé : Classification (ex. Random Forest ou Régression Logistique)

*Stratégie d'évaluation :*

- *Division des données* : Utiliser une séparation *train/validation/test* (typiquement 70/15/15 ou 80/10/10). La validation sert au tuning des hyperparamètres.
- *Validation croisée* : Appliquer une *k-fold cross-validation* (k=5 ou 10) ou *stratified k-fold* pour les classes déséquilibrées, afin d'obtenir une estimation robuste des performances.
- *Métriques principales* :
  - *Accuracy* (précision globale) : Utile quand les classes sont équilibrées.
  - *Precision, Recall, F1-score* : Essentielles en cas de déséquilibre de classes (ex. détection de fraudes ou maladies).
  - *AUC-ROC* et *courbe PR (Precision-Recall)* : Pour évaluer la capacité à discriminer les classes, surtout avec des seuils de décision variables.
  - *Matrice de confusion* : Pour analyser les faux positifs et faux négatifs.
  - Métriques additionnelles selon le contexte : Log-loss (pour probabilités), Cohen’s Kappa, etc.

*Méthodes complémentaires* :
- Courbes d’apprentissage (learning curves) pour détecter le sous-apprentissage ou surapprentissage.
- Analyse de feature importance (pour les modèles interprétables).
- Tests statistiques (ex. McNemar’s test) pour comparer deux modèles.

*Défis et limites* :
- *Déséquilibre de classes* : L’accuracy devient trompeuse → privilégier F1 ou AUC-PR.
- *Données bruitées ou mal étiquetées* : L’évaluation sur le test set peut surestimer les performances réelles.
- *Généralisation* : Risque de fuite de données (data leakage) si la séparation n’est pas temporelle ou par groupe.
- *Coût computationnel* : La validation croisée est coûteuse sur de très grands jeux de données.

 2. Modèle d'apprentissage non supervisé : Clustering (ex. K-Means, DBSCAN ou Hierarchical Clustering)

*Stratégie d'évaluation :*

L’évaluation est plus difficile car il n’y a pas de « vérité terrain » (ground truth) explicite.

*Métriques internes (sans labels)* :
- *Score de silhouette* : Mesure la cohésion intra-cluster et la séparation inter-clusters (valeur entre -1 et 1).
- *Méthode du coude (Elbow Method)* : Pour déterminer le nombre optimal de clusters K en regardant l’inertie (within-cluster sum of squares).
- *Indice de Calinski-Harabasz* et *Indice de Davies-Bouldin* : Évaluent le ratio variance inter/intra-clusters.
- *Indice de Dunn* : Rapport entre la distance minimale inter-clusters et la taille maximale intra-cluster.

*Métriques externes (si labels partiels disponibles)* :
- Adjusted Rand Index (ARI), Normalized Mutual Information (NMI), Homogeneity, Completeness, V-measure.

*Approches pratiques* :
- Visualisation : t-SNE, UMAP ou PCA + plots 2D/3D.
- Validation de stabilité : Ré-exécuter le clustering sur des sous-échantillons et mesurer la similarité des partitions.
- Utilité métier : Mesurer l’impact downstream (ex. amélioration d’un modèle supervisé utilisant les clusters comme features).

*Défis et limites* :
- *Subjectivité* : Plusieurs partitions peuvent être valides selon le niveau de granularité souhaité.
- *Sensibilité aux hyperparamètres* : K-Means est très sensible à l’initialisation et aux outliers.
- *Données de haute dimension* : La malédiction de la dimensionnalité rend les distances moins significatives.
- *Absence de métrique universelle* : Il faut souvent combiner plusieurs indicateurs + expertise du domaine.
- *Scalabilité* : Les mesures comme le silhouette score sont coûteuses sur de très gros jeux de données.

3. Modèle d'apprentissage par renforcement (ex. Q-Learning, DQN, PPO)

*Stratégie d'évaluation :*

L’évaluation se fait principalement pendant et après l’entraînement dans l’environnement.

*Métriques principales* :
- *Récompense cumulative moyenne* (ou discounted return) : Moyenne sur plusieurs épisodes.
- *Taux de succès* : Pourcentage d’épisodes où l’agent atteint l’objectif (ex. gagner une partie, atteindre un score cible).
- *Convergence* : Suivi de la courbe d’apprentissage (récompense en fonction des épisodes ou steps). Détecter un plateau.
- *Stabilité* : Variance de la performance sur plusieurs seeds d’entraînement (reproductibilité).
- *Équilibre exploration/exploitation* : Suivi de l’epsilon (dans ε-greedy) ou des métriques d’entropie de la politique. Trop d’exploration = apprentissage lent ; trop d’exploitation = risque de convergence locale.

*Méthodes complémentaires* :
- *Evaluation hold-out* : Tester l’agent entraîné sur des environnements légèrement différents (robustesse, généralisation).
- *Ablation studies* : Retirer des composants (ex. replay buffer, target network) pour mesurer leur impact.
- *Benchmark* : Comparaison avec des baselines (random agent, heuristiques) et avec l’état de l’art sur des environnements standards (Gym, Atari, MuJoCo).
- *Analyse qualitative* : Visualisation des trajectoires, heatmaps d’actions, etc.

*Défis et limites* :
- *Variance élevée* : Les performances peuvent fluctuer énormément d’un run à l’autre → nécessite de nombreuses seeds.
- *Signal de récompense sparse/delayed* : Difficile de savoir si l’agent progresse vraiment.
- *Coût computationnel* : L’entraînement est souvent très long et nécessite beaucoup de simulations.
- *Généralisation* : Un agent performant dans un environnement simulé peut échouer dans le monde réel (sim-to-real gap).
- *Sécurité et alignement* : Difficile d’évaluer les comportements émergents indésirables (ex. hacking de la récompense).
- *Exploration insuffisante* : L’agent peut ignorer des stratégies optimales rares.

               Synthèse et bonnes pratiques globales

| Catégorie          | Métrique principale          | Méthode clé                  | Principal défi                  |
|--------------------|------------------------------|------------------------------|---------------------------------|
| Supervisé          | F1-score / AUC-ROC           | Cross-validation             | Déséquilibre & data leakage     |
| Non supervisé      | Silhouette / Elbow           | Visualisation + stabilité    | Subjectivité & manque de ground truth |
| Renforcement       | Récompense cumulative        | Courbes d’apprentissage      | Variance & coût computationnel  |

*Conseils transversaux* :
- Toujours utiliser plusieurs métriques complémentaires.
- Privilégier l’évaluation sur des données/épisodes non vus.
- Documenter les conditions expérimentales (seeds, hyperparamètres, versions d’environnements).
- Considérer l’impact métier réel au-delà des métriques techniques.
