## Classe `Circle` avec propriétés et méthodes Dunder

Nous allons créer la classe `Circle` en implémentant les fonctionnalités demandées : initialisation par rayon ou diamètre, calcul de l'aire, représentation textuelle, comparaison et addition de cercles, et tri.

```
import math
from functools import total_ordering

@total_ordering
class Circle:
    def __init__(self, radius):
        if radius < 0:
            raise ValueError("Le rayon ne peut pas être négatif.")
        self._radius = radius

    @classmethod
    def from_diameter(cls, diameter):
        if diameter < 0:
            raise ValueError("Le diamètre ne peut pas être négatif.")
        return cls(diameter / 2)

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Le rayon ne peut pas être négatif.")
        self._radius = value

    @property
    def diameter(self):
        return self._radius * 2

    @diameter.setter
    def diameter(self, value):
        if value < 0:
            raise ValueError("Le diamètre ne peut pas être négatif.")
        self._radius = value / 2

    def area(self):
        return math.pi * (self._radius ** 2)

    def __str__(self):
        return f"Cercle de rayon {self.radius:.2f} et de diamètre {self.diameter:.2f}"

    def __repr__(self):
        return f"Circle(radius={self.radius})"

    def __add__(self, other):
        if isinstance(other, Circle):
            new_radius = self.radius + other.radius
            return Circle(new_radius)
        elif isinstance(other, (int, float)):
            new_radius = self.radius + other
            if new_radius < 0:
                raise ValueError("L'addition ne peut pas résulter en un rayon négatif.")
            return Circle(new_radius)
        else:
            return NotImplemented

    def __eq__(self, other):
        if isinstance(other, Circle):
            return self.radius == other.radius
        return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Circle):
            return self.radius < other.radius
        return NotImplemented


# --- Démonstration ---
```

### Création d'instances et accès aux propriétés

```
c1 = Circle(radius=5)
c2 = Circle.from_diameter(12)

print(f"Cercle c1: {c1}")
print(f"Rayon de c1: {c1.radius}")
print(f"Diamètre de c1: {c1.diameter}")
print(f"Aire de c1: {c1.area():.2f}")

print(f"\nCercle c2: {c2}")
print(f"Rayon de c2: {c2.radius}")
print(f"Diamètre de c2: {c2.diameter}")
print(f"Aire de c2: {c2.area():.2f}")
```

### Modification des propriétés

```
c1.radius = 10
print(f"Cercle c1 après modification du rayon: {c1}")
print(f"Aire de c1: {c1.area():.2f}")

c2.diameter = 20
print(f"Cercle c2 après modification du diamètre: {c2}")
print(f"Aire de c2: {c2.area():.2f}")
```

### Addition de cercles

```
c3 = c1 + c2
print(f"Cercle c3 (c1 + c2): {c3}")
print(f"Aire de c3: {c3.area():.2f}")

c4 = c1 + 3
print(f"Cercle c4 (c1 + 3): {c4}")
```

### Comparaison de cercles

```
c5 = Circle(radius=7)
c6 = Circle(radius=10)
c7 = Circle(radius=7)

print(f"c5 ({c5.radius}) == c7 ({c7.radius}): {c5 == c7}")
print(f"c5 ({c5.radius}) > c6 ({c6.radius}): {c5 > c6}")
print(f"c6 ({c6.radius}) > c5 ({c5.radius}): {c6 > c5}")
print(f"c5 ({c5.radius}) < c6 ({c6.radius}): {c5 < c6}")
print(f"c6 ({c6.radius}) < c5 ({c5.radius}): {c6 < c5}")
```

### Tri d'une liste de cercles

```
circles = [Circle(radius=3), Circle(radius=1), Circle(radius=8), Circle(radius=5)]
print(f"Liste de cercles non triée: {circles}")

sorted_circles = sorted(circles)
print(f"Liste de cercles triée: {sorted_circles}")
```
