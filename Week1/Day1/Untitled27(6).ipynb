Exercice 7 : Pair ou impair

````
# Demonstration de la logique pair/impair avec l'opérateur modulo

# Exemple 1: Nombre pair
number1 = 10
result1 = number1 % 2
status1 = "pair" if result1 == 0 else "impair"
print(f"{number1} % 2 = {result1} \u2192 {status1} ({number1} = 2 \u00d7 {number1 // 2}, reste {result1})")

# Exemple 2: Nombre impair
number2 = 7
result2 = number2 % 2
status2 = "pair" if result2 == 0 else "impair"
print(f"{number2} % 2 = {result2} \u2192 {status2} ({number2} = 2 \u00d7 {number2 // 2}, reste {result2})")
````

## Comment installer des bibliothèques Python ?

Pour installer des bibliothèques Python, vous utilisez généralement `pip`, le gestionnaire de paquets par défaut pour Python. Voici la commande de base :

```bash
pip install nom_de_la_bibliothèque
```

Si vous souhaitez installer une version spécifique, vous pouvez le faire comme ceci :

```bash
pip install nom_de_la_bibliothèque==version_specifique
```

Si vous êtes dans un environnement Jupyter ou Colab, vous devez souvent préfixer la commande `pip` avec un point d'exclamation `!` pour l'exécuter comme une commande shell :

```python
!pip install nom_de_la_bibliothèque
```

Par exemple, pour installer la bibliothèque `requests` (souvent utilisée pour faire des requêtes HTTP) :

````
# Exemple d'installation de la bibliothèque 'requests'
!pip install requests

# Vous pouvez ensuite importer et utiliser la bibliothèque
import requests
print('La bibliothèque requests est maintenant installée et peut être importée.')
````
