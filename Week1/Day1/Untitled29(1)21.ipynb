```python
# Demonstration for Challenge 2: Remove Consecutive Duplicates

def remove_consecutive_duplicates(word):
    result = []
    for char in word:
        if not result or char != result[-1]:
            result.append(char)
    return "".join(result)

demo_word = "aaabbbcccdddeeefff"
demo_cleaned_word = remove_consecutive_duplicates(demo_word)
print(f"Demonstration: Original word → \"{demo_word}\"")
print(f"Demonstration: Cleaned word → \"{demo_cleaned_word}\"")
```
