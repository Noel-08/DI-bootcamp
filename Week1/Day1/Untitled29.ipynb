```python
# ─────────────────────────────────────────
#  Challenge 1 – List of Multiples
# ─────────────────────────────────────────

def get_multiples(number, length):
    return [number * i for i in range(1, length + 1)]

print("=== Challenge 1 : Multiples ===")
number = int(input("Enter a number  : "))
length = int(input("Enter the length: "))

multiples = get_multiples(number, length)
print(f"Multiples of {number} (length {length}) → {multiples}")


# ─────────────────────────────────────────
#  Challenge 2 – Remove Consecutive Duplicates
# ─────────────────────────────────────────

def remove_consecutive_duplicates(word):
    result = []
    for char in word:
        if not result or char != result[-1]:
            result.append(char)
    return "".join(result)

print("\n=== Challenge 2 : Remove Consecutive Duplicates ===")
word = input("Enter a word: ")

cleaned = remove_consecutive_duplicates(word)
print(f"Cleaned word → \"{cleaned}\"")
```
