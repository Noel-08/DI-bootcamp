### Étape 1 : Créer la classe Chien

```python
class Dog:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        return f'{self.name} aboie'

    def run_speed(self):
        return (self.weight / self.age) * 10

    def fight(self, other_dog):
        self_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight

        if self_power > other_power:
            return f'{self.name} a gagné le combat contre {other_dog.name}!'
        elif other_power > self_power:
            return f'{other_dog.name} a gagné le combat contre {self.name}!'
        else:
            return f'Le combat entre {self.name} et {other_dog.name} est une égalité!'
```

### Étape 2 : Créer des instances de chien

```python
dog1 = Dog('Buddy', 5, 25) # Name, Age, Weight
dog2 = Dog('Daisy', 3, 18)
dog3 = Dog('Max', 7, 30)
```

### Étape 3 : Méthodes de test sur les chiens

```python
# Test bark() method
print(f'{dog1.name}: {dog1.bark()}')
print(f'{dog2.name}: {dog2.bark()}')

# Test run_speed() method
print(f'{dog1.name} run speed: {dog1.run_speed():.2f}')
print(f'{dog2.name} run speed: {dog2.run_speed():.2f}')
print(f'{dog3.name} run speed: {dog3.run_speed():.2f}')

# Test fight() method
print(dog1.fight(dog2))
print(dog2.fight(dog3))
print(dog3.fight(dog1))
```

### Étape 1 & 2 : Créer la classe `PetDog` (avec la classe `Dog` déjà définie)

```python
import random

class PetDog(Dog):
    def __init__(self, name, age, weight):
        super().__init__(name, age, weight)
        self.trained = False

    def train(self):
        print(self.bark())
        self.trained = True

    def play(self, *args):
        dog_names = [self.name] + [dog.name for dog in args]
        print(f'{', '.join(dog_names)} tous jouent ensemble')

    def do_a_trick(self):
        if self.trained:
            tricks = ["does a barrel roll", "stands on his back legs", "shakes your hand", "plays dead"]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} n'est pas dressé pour faire des tours.")
```

### Étape 3 : Tester les méthodes `PetDog`

```python
my_pet_dog = PetDog('Roxy', 4, 20)
other_pet_dog = PetDog('Spot', 2, 15)

print(f"Initial trained status for {my_pet_dog.name}: {my_pet_dog.trained}")

# Test do_a_trick() before training
my_pet_dog.do_a_trick()

# Test train() method
my_pet_dog.train()
print(f"Trained status for {my_pet_dog.name} after training: {my_pet_dog.trained}")

# Test do_a_trick() after training
my_pet_dog.do_a_trick()
my_pet_dog.do_a_trick()

# Test play() method
my_pet_dog.play(other_pet_dog, dog1) # dog1 is from the previous exercise
```

### Étape 1 : Créer la `Person` classe

```python
class Person:
    def __init__(self, first_name, age, last_name=""):
        self.first_name = first_name
        self.age = age
        self.last_name = last_name

    def is_18(self):
        return self.age >= 18
```

### Étape 2 : Créer la `Family` classe

```python
class Family:
    def __init__(self, last_name):
        self.last_name = last_name
        self.members = []

    def born(self, first_name, age):
        person = Person(first_name, age, self.last_name)
        self.members.append(person)
        print(f'{first_name} {self.last_name} est né dans la famille.')

    def check_majority(self, first_name):
        found_person = None
        for member in self.members:
            if member.first_name == first_name:
                found_person = member
                break

        if found_person:
            if found_person.is_18():
                print(f"You are over 18, your parents Jane and John accept that you will go out with your friends")
            else:
                print(f"Sorry, you are not allowed to go out with your friends.")
        else:
            print(f"La personne avec le prénom {first_name} n'a pas été trouvée dans la famille.")

    def family_presentation(self):
        print(f"Nom de famille: {self.last_name}")
        print("Membres de la famille:")
        for member in self.members:
            print(f"- {member.first_name} (âge: {member.age})")
```

### Comportement attendu (Démonstration)

```python
# Créez une famille
dupont_family = Family('Dupont')

# Ajoutez des membres
dupont_family.born('Alice', 20)
dupont_family.born('Bob', 15)
dupont_family.born('Carla', 10)
dupont_family.born('David', 30)

print('\n--- Présentation de la famille ---\n')
dupont_family.family_presentation()

print('\n--- Vérification de la majorité ---\n')
dupont_family.check_majority('Alice')
dupont_family.check_majority('Bob')
dupont_family.check_majority('Carla')
dupont_family.check_majority('John') # Personne non existante
```
