Exercice 2 : Statistiques descriptives

```python
import numpy as np
from scipy import stats

# Sample data provided by the user
data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calculate descriptive statistics using SciPy and NumPy
mean_val = np.mean(data)
median_val = np.median(data)
variance_val = np.var(data, ddof=1) # ddof=1 for sample variance
std_dev_val = np.std(data, ddof=1) # ddof=1 for sample standard deviation

print(f"Data: {data}")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")
```

Exercice 3 : Comprendre les distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 50  # mean
sigma = 10  # standard deviation

# Generate a sample from the normal distribution
sample_size = 1000
np.random.seed(42) # for reproducibility
data_normal = norm.rvs(loc=mu, scale=sigma, size=sample_size)

# Create a figure and an axes object for the plot
plt.figure(figsize=(10, 6))

# Plot the histogram of the generated data
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='g', label='Histogram of Sample Data')

# Plot the Probability Density Function (PDF)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Normal PDF')

plt.title('Normal Distribution (Mean=50, Std Dev=10)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

Exercice 4 : Application du test T

```python
import numpy as np
from scipy import stats

# Generate two independent samples
np.random.seed(42) # for reproducibility
data1 = np.random.normal(50, 10, 100) # mean=50, std=10, size=100
data2 = np.random.normal(60, 10, 100) # mean=60, std=10, size=100

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(data1, data2)

print(f"Data 1 (Mean): {np.mean(data1):.2f}")
print(f"Data 2 (Mean): {np.mean(data2):.2f}")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("\nConclusion: There is a statistically significant difference between the means of the two datasets.")
else:
    print("\nConclusion: There is no statistically significant difference between the means of the two datasets.")
```

 Exercice 5 : Analyse de régression linéaire

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datasets provided by the user
house_sizes = np.array([50, 70, 80, 100, 120])  # in square meters
house_prices = np.array([150000, 200000, 210000, 250000, 280000]) # in monetary units

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# Question 1: What is the slope and the intercept of the regression line?
print(f"Slope (Pente): {slope:.2f}")
print(f"Intercept (Intersection): {intercept:.2f}")
```

```python
# Question 2: Predict the price of a 90 square meter house.
predicted_price_90sqm = slope * 90 + intercept
print(f"Predicted price for a 90 square meter house: {predicted_price_90sqm:.2f}")
```

### Interprétation de la pente:

La pente de la droite de régression représente le changement moyen dans le prix de la maison pour chaque augmentation d'une unité (dans ce cas, un mètre carré) de la taille de la maison.

Dans le contexte des prix de l'immobilier, une pente de `{slope:.2f}` signifie que pour chaque mètre carré supplémentaire, le prix de la maison augmente en moyenne de `{slope:.2f}` unités monétaires, en supposant que toutes les autres variables restent constantes.

```python
# Optional: Visualize the regression line
plt.figure(figsize=(10, 6))
plt.scatter(house_sizes, house_prices, label='Données réelles')
plt.plot(house_sizes, slope * house_sizes + intercept, color='red', label='Ligne de régression linéaire')
plt.scatter([90], [predicted_price_90sqm], color='green', marker='x', s=100, label=f'Prédiction (90m²): {predicted_price_90sqm:.2f}')
plt.title('Régression Linéaire : Prix des Maisons vs Taille')
plt.xlabel('Taille de la maison (mètres carrés)')
plt.ylabel('Prix de la maison (unités monétaires)')
plt.legend()
plt.grid(True)
plt.show()
```

Exercice 6 : Comprendre l’ANOVA

```python
import numpy as np
from scipy import stats

# Datasets provided by the user
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

# Question: D’après la valeur P, les engrais ont-ils des effets significativement différents sur la croissance des plantes ?
alpha = 0.05
if p_value < alpha:
    print(f"\nConclusion (based on P-value {p_value:.3f} < {alpha}): The fertilizers have a statistically significant different effect on plant growth.")
else:
    print(f"\nConclusion (based on P-value {p_value:.3f} >= {alpha}): The fertilizers do not have a statistically significant different effect on plant growth.")
```

### Interprétation de la valeur P

**Explication si la valeur P était supérieure à 0,05 :**

Si la valeur P avait été supérieure à 0,05 (le seuil de signification commun, ou alpha), nous n'aurions pas pu rejeter l'hypothèse nule. L'hypothèse nulle en ANOVA est qu'il n'y a pas de différence significative entre les moyennes des groupes (dans ce cas, l'effet des différents engrais sur la croissance des plantes).

En d'autres termes, si P > 0,05, nous conclurions qu'il n'y a pas suffisamment de preuves statistiques pour affirmer que les engrais ont des effets significativement différents sur la croissance des plantes. Toute différence observée entre les groupes serait probablement due au hasard plutôt qu'à l'efficacité de l'engrais lui-même.

Exercice 7 : Distributions de probabilité (optionnelle)

```python
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials (coin tosses)
k = 5   # Number of successes (heads)
p = 0.5 # Probability of success on each trial (fair coin)

# Calculate the probability mass function (PMF)
probability = binom.pmf(k, n, p)

print(f"The probability of getting exactly {k} heads in {n} tosses is: {probability:.4f}")
```

Exercice 8 : Coefficients de corrélation (optionnel)


```python
import pandas as pd
from scipy import stats

# Dataset provided by the user
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

print("Original Data:")
display(data)

# Calculate Pearson correlation coefficient
pearson_corr, pearson_pvalue = stats.pearsonr(data['age'], data['income'])
print(f"\nPearson Correlation Coefficient: {pearson_corr:.3f} (P-value: {pearson_pvalue:.3f})")

# Calculate Spearman correlation coefficient
spearman_corr, spearman_pvalue = stats.spearmanr(data['age'], data['income'])
print(f"Spearman Correlation Coefficient: {spearman_corr:.3f} (P-value: {spearman_pvalue:.3f})")
```
