Exercise 6: Identifying Data Types


### Exercise 6: Identifying Data Types - Structured vs. Unstructured

To categorize these data sources, let's first define **structured** and **unstructured** data:

*   **Structured Data:** Data that has a predefined format and is organized in a fixed schema, typically tabular. It's easy to store, query, and analyze using conventional databases (e.g., relational databases).
    *   **Examples:** Database tables, Excel spreadsheets, CSV files with consistent columns.

*   **Unstructured Data:** Data that does not have a predefined format or organization. It can be textual or non-textual and often requires more advanced techniques (like natural language processing or machine learning) for analysis.
    *   **Examples:** Text documents, images, audio, video, social media posts.

---

#### Classification of Provided Data Sources:

1.  **A company’s financial reports stored in an Excel file.**
    *   **Classification:** Structured Data
    *   **Justification:** Excel files typically organize financial data into rows and columns with clear headers and data types (e.g., dates, numbers, text), fitting a predefined tabular schema.

2.  **Photographs uploaded to a social media platform.**
    *   **Classification:** Unstructured Data
    *   **Justification:** Photographs are visual media. While they might have metadata (like date, location, camera type), the core content of the image itself (pixels, objects, scenes) does not conform to a fixed, tabular structure.

3.  **A collection of news articles on a website.**
    *   **Classification:** Unstructured Data
    *   **Justification:** News articles are primarily text. Although they might have some structured elements like publication date or author, the main body of the text content is free-form and does not fit into a rigid, predefined data model.

4.  **Inventory data in a relational database.**
    *   **Classification:** Structured Data
    *   **Justification:** Relational databases are designed specifically to store and manage structured data. Inventory data would be organized into tables with columns for items, quantities, prices, locations, etc., all adhering to a strict schema.

5.  **Recorded interviews from a market research study.**
    *   **Classification:** Unstructured Data
    *   **Justification:** Recorded interviews are audio (or video) files. The content is spoken language, which is free-form and lacks a predefined tabular structure. Analyzing this would typically involve transcription and natural language processing.

Exercise 7: Transformation Exercise


### Exercise 7: Transformation Exercise - Converting Unstructured to Structured Data

This exercise focuses on understanding how various unstructured data sources can be transformed into structured formats, making them amenable to quantitative analysis and database storage.

---

#### Methods for Converting Unstructured Data to Structured Data:

1.  **A series of blog posts about travel experiences.**
    *   **Method:**
        1.  **Text Extraction & Cleaning:** Extract the raw text from each blog post. Remove HTML tags, advertisements, and irrelevant boilerplate text.
        2.  **Natural Language Processing (NLP):** Apply NLP techniques such as:
            *   **Named Entity Recognition (NER):** Identify and extract entities like locations (cities, countries), attractions, dates, and names of people/companies mentioned.
            *   **Sentiment Analysis:** Determine the emotional tone (positive, negative, neutral) of the travel experiences.
            *   **Topic Modeling:** Identify main themes or topics discussed in each post (e.g., 'adventure travel', 'beach holidays', 'cultural immersion').
            *   **Keyword Extraction:** Identify relevant keywords and phrases.
        3.  **Feature Extraction:** Extract quantitative features such as word count, reading time, number of images, number of comments.
    *   **Structured Output Examples:** A table with columns like `Post_ID`, `Title`, `Author`, `Publish_Date`, `Main_Location`, `Sentiment_Score`, `Dominant_Topic`, `Keywords`, `Word_Count`, `Avg_Rating` (if available).
    *   **Reasoning:** Blog posts are free-form text. NLP is essential to extract meaningful entities and sentiments, turning qualitative observations into quantifiable and categorizable data. This allows for analysis like 'What are the most positively reviewed travel destinations?', 'Which topics are most frequently discussed by travelers?', or 'How does post length correlate with engagement?'

2.  **Audio recordings of customer service calls.**
    *   **Method:**
        1.  **Speech-to-Text (STT) Transcription:** Convert the audio recordings into written transcripts.
        2.  **NLP on Transcripts:** Similar to blog posts, apply NLP:
            *   **Sentiment Analysis:** Assess customer satisfaction or frustration during the call.
            *   **Intent Recognition:** Identify the customer's purpose for calling (e.g., 'technical support', 'billing inquiry', 'product complaint').
            *   **Keyword/Phrase Detection:** Identify specific product names, common complaints, or resolution steps.
            *   **Entity Extraction:** Extract customer names, product IDs, order numbers.
        3.  **Acoustic Feature Extraction (Optional but valuable):** Analyze non-linguistic cues from the audio like pitch, volume, speaking rate, which can indicate emotional states.
    *   **Structured Output Examples:** A table with columns like `Call_ID`, `Date_Time`, `Customer_ID`, `Agent_ID`, `Call_Duration`, `Customer_Sentiment`, `Call_Intent`, `Keywords_Detected`, `Issue_Resolution_Status`.
    *   **Reasoning:** Audio is inherently unstructured. STT converts it to text, and then NLP can extract critical information about customer interactions, issues, and sentiment. This enables analysis such as 'What are the most common reasons for negative sentiment calls?', 'Which agents have the highest resolution rates for technical issues?', or 'How does call duration relate to customer satisfaction?'

3.  **Handwritten notes from a brainstorming session.**
    *   **Method:**
        1.  **Optical Character Recognition (OCR):** Use OCR technology to convert handwritten text into digital, machine-readable text.
        2.  **Keyword and Idea Extraction:** Manually or semi-automatically extract key ideas, concepts, and action items from the digitized text.
        3.  **Categorization/Tagging:** Group related ideas into categories or assign tags (e.g., 'marketing', 'product feature', 'risk', 'solution').
        4.  **Relationship Mapping:** Identify connections between ideas (e.g., 'idea X supports idea Y').
    *   **Structured Output Examples:** A table with columns like `Note_ID`, `Session_Date`, `Participant`, `Idea_Text`, `Category`, `Keywords`, `Related_Ideas`, `Action_Item` (boolean), `Priority`.
    *   **Reasoning:** Handwritten notes are highly unstructured. OCR provides the initial digitization. Subsequent manual or NLP-assisted extraction and categorization are needed to structure the qualitative content into actionable data. This helps in tracking brainstorming output, identifying key themes, prioritizing ideas, and assigning tasks.

4.  **A video tutorial on cooking.**
    *   **Method:**
        1.  **Video Segmentation:** Break down the video into logical segments (e.g., 'introduction', 'ingredient preparation', 'cooking step 1', 'plating').
        2.  **Speech-to-Text (STT) & NLP:** Transcribe spoken instructions and dialogue. Use NLP to extract:
            *   **Ingredients:** List all ingredients mentioned.
            *   **Cooking Steps:** Detail each instruction (e.g., 'chop onions', 'sauté for 5 minutes', 'bake at 350°F').
            *   **Tools/Equipment:** Identify kitchen tools used.
            *   **Time Durations:** Extract cooking times, prep times.
        3.  **Object Recognition (Computer Vision):** Use computer vision to identify objects in the video frames (e.g., 'knife', 'pot', 'carrot', 'finished dish').
        4.  **Action Recognition:** Identify cooking actions being performed (e.g., 'mixing', 'slicing', 'stirring').
        5.  **Metadata Extraction:** Extract video metadata like title, description, tags, upload date.
    *   **Structured Output Examples:** A table with columns like `Video_ID`, `Recipe_Name`, `Cuisine_Type`, `Total_Prep_Time`, `Total_Cook_Time`, `Key_Ingredients`, `Tools_Used`, `Complexity_Level`, `Step_by_Step_Instructions` (as a list or structured text), `Sentiment_Score` (from narration).
    *   **Reasoning:** Video is a rich, complex unstructured data type. Combining STT, NLP, and computer vision allows for comprehensive extraction of both auditory and visual information. This enables structured analysis such as 'What are the most common ingredients in quick recipes?', 'Which cooking techniques are frequently demonstrated?', or 'How long does an average recipe tutorial last?'

Exercise 9 : Export a dataframe to excel format and JSON format.

```python
import pandas as pd

# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df_export = pd.DataFrame(data)

print("Original DataFrame:")
display(df_export)
```

```python
# Export the DataFrame to an Excel file
excel_file_name = 'exported_dataframe.xlsx'
df_export.to_excel(excel_file_name, index=False)
print(f"DataFrame exported to {excel_file_name}")
```

```python
# Export the DataFrame to a JSON file
json_file_name = 'exported_dataframe.json'
df_export.to_json(json_file_name, orient='records', indent=4)
print(f"DataFrame exported to {json_file_name}")
```
