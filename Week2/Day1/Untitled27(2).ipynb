Introduction

Qu’est-ce que l’analyse des données ?

Cette discipline utilise plusieurs outils technologiques tels que Python, Pandas, Matplotlib, SQL ou encore les plateformes de visualisation de données. Grâce à ces outils, les analystes peuvent traiter efficacement de grandes quantités d’informations et produire des rapports fiables.

L’objectif principal de l’analyse des données est d’aider les organisations à prendre des décisions basées sur des faits concrets plutôt que sur de simples suppositions.

Exercice 3 : Identification des Types de Données

Introduction

Dans l’analyse des données, l’identification correcte des types de données constitue une étape essentielle avant toute analyse statistique ou modélisation. Les données peuvent être classées en deux grandes catégories : les données quantitatives et les données qualitatives.

Les données quantitatives représentent des valeurs numériques mesurables pouvant être utilisées pour effectuer des calculs mathématiques ou statistiques.

Les données qualitatives représentent des caractéristiques descriptives ou catégorielles qui permettent de classifier ou décrire des individus, des objets ou des événements.


Cette classification facilite le choix des méthodes d’analyse, des visualisations et des techniques de traitement appropriées.


---

En revanche, la colonne Species est qualitative car elle sert uniquement à identifier une catégorie ou une classe de fleur. Elle ne possède pas de valeur numérique mesurable.

Certaines colonnes numériques comme PassengerId ou Pclass sont considérées qualitatives car elles servent à identifier ou classer les individus plutôt qu’à mesurer une quantité réelle.



---

Importance de l’identification des types de données

L’identification correcte des types de données présente plusieurs avantages :

1. Choix des méthodes statistiques adaptées

Les données quantitatives peuvent être analysées avec des moyennes, médianes ou corrélations, tandis que les données qualitatives utilisent généralement des fréquences ou des pourcentages.

2. Sélection des visualisations appropriées

Les données quantitatives sont souvent représentées par des histogrammes ou des courbes.

Les données qualitatives sont mieux représentées par des diagrammes à barres ou des graphiques circulaires.


3. Préparation des données pour le Machine Learning

Les algorithmes de machine learning nécessitent souvent la conversion des données qualitatives en valeurs numériques à travers des techniques comme le Label Encoding ou le One-Hot Encoding.


---

Conclusion

En conclusion, la classification des données en variables quantitatives et qualitatives constitue une étape fondamentale dans tout projet d’analyse de données. Cette étape permet de mieux comprendre la structure des datasets et d’appliquer les méthodes d’analyse les plus pertinentes.

Les datasets Iris et Titanic illustrent parfaitement la coexistence de différents types de données dans un même ensemble d’informations. La maîtrise de cette distinction est donc essentielle pour produire des analyses fiables, pertinentes et professionnelles.

 Exercice 6 : Identification des types de données (Structurées vs Non structurées)

### Définition rapide :
- *Données structurées* : Données organisées selon un format prédéfini (tableaux, colonnes, lignes) faciles à exploiter par un ordinateur (ex. : bases de données, Excel).
- *Données non structurées* : Données sans format fixe, souvent libres (texte libre, images, audio, vidéos).

---

### Analyse des sources de données :

| N° | Source de données | Type de données          | Explication |
|----|--------------------|---------------------------|-----------|
| 1  | Les rapports financiers d'une entreprise sont stockés dans un fichier Excel. | *Structurées* | Les données sont organisées en lignes et colonnes avec des champs définis (chiffre d'affaires, dépenses, bénéfices, dates, etc.). Excel impose une structure tabulaire. |
| 2  | Photographies téléchargées sur une plateforme de médias sociaux. | *Non structurées* | Ce sont des images. Il n’y a pas de format tabulaire. Les pixels ne suivent pas une structure prédéfinie exploitable directement. |
| 3  | Une collection d'articles de presse sur un site web. | *Non structurées* | Le contenu principal est du texte libre (paragraphes, langage naturel). Même si le site web a une structure HTML, le contenu des articles reste non structuré. |
| 4  | Données d'inventaire dans une base de données relationnelle. | *Structurées* | Les bases de données relationnelles (ex. : MySQL, PostgreSQL) utilisent des tables avec des colonnes bien définies (Produit, Quantité, Prix, Fournisseur, etc.). |
| 5  | Enregistrements d'entretiens issus d'une étude de marché. | *Non structurées* | Il s’agit généralement d’enregistrements audio ou de transcriptions en texte libre. Les réponses ne suivent pas un format fixe et prédéfini. |


Développement détaillé de l’Exercice 6*

Voici une explication plus approfondie pour chaque source de données, avec des exemples concrets de traitement possibles.

### 1. Les rapports financiers d'une entreprise stockés dans un fichier Excel
- *Type* : *Données structurées*
- *Explication* : Les données sont organisées dans un tableau avec des colonnes clairement définies (ex. : Date, Chiffre d’affaires, Dépenses, Bénéfice net, etc.). Chaque ligne représente un enregistrement cohérent.
- *Exemples de traitement* :
  - Analyse avec des formules Excel ou Power BI
  - Calcul de moyennes, tendances, ratios financiers
  - Importation directe dans Python (pandas) ou R pour des analyses avancées

### 2. Photographies téléchargées sur une plateforme de médias sociaux
- *Type* : *Données non structurées*
- *Explication* : Ce sont des images binaires. Il n’existe pas de colonnes ou de champs prédéfinis. L’information est contenue dans les pixels.
- *Exemples de traitement* :
  - Utilisation de l’Intelligence Artificielle (Computer Vision) : reconnaissance d’objets, de visages, de sentiments
  - Extraction de métadonnées (date, localisation, appareil photo)
  - Stockage dans un système de gestion d’images ou une base NoSQL

### 3. Une collection d'articles de presse sur un site web
- *Type* : *Données non structurées*
- *Explication* : Le contenu principal est du texte libre (titres, paragraphes, langage naturel). Même si la page web a une structure HTML, le texte des articles n’est pas organisé en colonnes.
- *Exemples de traitement* :
  - *Text Mining* / NLP : analyse de sentiments, extraction de mots-clés, résumé automatique
  - Scraping pour extraire le texte puis le stocker dans une base de données vectorielle
  - Classification thématique (politique, économie, sport…)

### 4. Données d'inventaire dans une base de données relationnelle
- *Type* : *Données structurées*
- *Explication* : Les données sont stockées dans des tables avec des relations (ex. : table Produits, table Stocks, table Fournisseurs) et des schémas bien définis.
- *Exemples de traitement* :
  - Requêtes SQL pour générer des rapports (ex. : produits en rupture de stock)
  - Analyse en temps réel des niveaux d’inventaire
  - Intégration avec des outils de Business Intelligence (Tableau, Power BI)

### 5. Enregistrements d'entretiens issus d'une étude de marché
- *Type* : *Données non structurées*
- *Explication* : Il s’agit d’audio ou de transcriptions en texte libre. Les réponses des personnes interrogées sont variées et ne respectent pas un format fixe.
- *Exemples de traitement* :
  - Transcription automatique (ex. : Whisper d’OpenAI)
  - Analyse de sentiments et d’opinions
  - Codage thématique manuel ou automatique pour identifier les tendances

---

*Résumé final :*

| Type de données     | Exemples                          | Difficulté d’analyse |
|---------------------|-----------------------------------|----------------------|
| *Structurées*     | Excel, Base de données            | Facile               |
| *Non structurées* | Photos, Articles, Entretiens      | Plus complexe        |

Les données *structurées* sont plus faciles à analyser directement, tandis que les données *non structurées* nécessitent souvent des techniques d’IA (Machine Learning, Deep Learning) pour en extraire de la valeur.

Exercice 7 : Transformation de données non structurées en données structurées*

Voici une proposition claire et réaliste pour chacune des sources de données.

 1. Une série d'articles de blog sur les expériences de voyage
*Méthode proposée :*  
Utiliser le *Natural Language Processing (NLP)* pour extraire les informations clés.

*Étapes :*
- Scraping des articles (récupération du texte)
- Application de modèles NLP (ex. : spaCy, Hugging Face) pour extraire :
  - Destination
  - Dates de voyage
  - Note globale / Satisfaction
  - Thèmes (plage, montagne, culture, gastronomie…)
  - Budget approximatif
  - Points positifs / négatifs (analyse de sentiments)

*Justification :*  
Les articles sont du texte libre. Le NLP permet de transformer ce texte en un tableau structuré (ex. : une ligne par voyage avec des colonnes claires). C’est une méthode automatisée, scalable et précise pour ce type de contenu narratif.

2. Enregistrements audio des appels au service client
*Méthode proposée :*  
*Transcription automatique + Analyse de texte*

*Étapes :*
- Transcription avec un outil comme *Whisper (OpenAI), **Google Speech-to-Text* ou *Deepgram*
- Analyse du texte transcrit avec NLP pour extraire :
  - Nom du client / Numéro d’appel
  - Date et durée de l’appel
  - Type de problème (technique, facturation, livraison…)
  - Sentiment du client (insatisfait, neutre, satisfait)
  - Résolution (oui/non) + Temps de résolution
  - Mots-clés récurrents

*Justification :*  
L’audio est impossible à analyser directement par une machine. La transcription le transforme en texte, puis le NLP structure l’information. C’est la méthode standard utilisée par la plupart des centres d’appels modernes pour créer des tableaux exploitables.

3. Notes manuscrites issues d'une séance de brainstorming
*Méthode proposée :*  
*OCR (Reconnaissance Optique de Caractères) + Structuration manuelle ou semi-automatique*

*Étapes :*
- Numérisation des notes (scan ou photo)
- Utilisation d’outils OCR (Google Vision, Microsoft Azure OCR, ou Tesseract)
- Nettoyage du texte obtenu
- Structuration :
  - Catégorisation manuelle ou avec LLM (ex. : Idée principale / Avantage / Inconvénient / Responsable / Priorité)
  - Création d’un tableau (Idée | Thème | Priorité | Participants | Statut)

*Justification :*  
Les notes manuscrites sont à la fois non structurées et souvent peu lisibles. L’OCR est indispensable pour passer du papier au numérique. Ensuite, une structuration humaine ou assistée par IA reste nécessaire car le brainstorming est souvent désorganisé par nature.

 4. Un tutoriel vidéo sur la cuisine
*Méthode proposée :*  
*Transcription vidéo + Extraction temporelle (Timestamping)*

*Étapes :*
- Transcription automatique avec Whisper ou YouTube Auto-captions
- Découpage du contenu par étapes grâce aux timestamps
- Extraction structurée avec IA :
  - Titre de la recette
  - Temps de préparation / Cuisson
  - Liste des ingrédients (avec quantités)
  - Étapes détaillées (numérotées)
  - Niveau de difficulté
  - Nombre de personnes
  - Astuces / Variantes

*Justification :*  
Une vidéo contient du son et des images. La transcription transforme la partie audio en texte, et l’ajout de timestamps permet de structurer le tutoriel étape par étape. Le résultat peut être stocké dans une base de données de recettes, très utile pour des applications culinaires.

*Résumé des méthodes :*

| Source de données                  | Méthode principale              | Niveau d’automatisation | Difficulté |
|------------------------------------|----------------------------------|--------------------------|----------|
| Articles de blog                   | NLP / Text Mining               | Élevé                    | Moyen    |
| Enregistrements audio              | Transcription + NLP             | Élevé                    | Moyen    |
| Notes manuscrites                  | OCR + Structuration             | Moyen                    | Élevé    |
| Tutoriel vidéo                     | Transcription + Timestamping    | Élevé                    | Moyen    |

 Exercice 9 : Exporter un dataframe au format Excel et au format JSON.


```
import pandas as pd

# =====================================================================
# ÉTAPE 1 : CRÉATION D'UN DATAFRAME SIMPLE
# =====================================================================
# Création d'un dictionnaire représentant de fausses données d'utilisateurs
data = {
    'Nom': ['Alice', 'Bob', 'Charlie', 'David'],
    'Âge': [25, 30, 35, 40],
    'Ville': ['Paris', 'Lyon', 'Marseille', 'Lille'],
    'Score': [85.5, 92.0, 78.5, 88.0]
}

# Conversion du dictionnaire en DataFrame Pandas
df_simple = pd.DataFrame(data)

print("--- DataFrame créé avec succès ---")
print(df_simple)
print("\n")

# =====================================================================
# ÉTAPE 2 : EXPORTATION AU FORMAT EXCEL (.xlsx)
# =====================================================================
# L'argument index=False permet de ne pas enregistrer la colonne des numéros de ligne (0, 1, 2...)
fichier_excel = "donnees_exportees.xlsx"
df_simple.to_excel(fichier_excel, index=False)
print(f"✅ Le DataFrame a été exporté au format Excel sous le nom : '{fichier_excel}'")

# =====================================================================
# ÉTAPE 3 : EXPORTATION AU FORMAT JSON (.json)
# =====================================================================
# L'argument orient='records' structure le JSON sous forme de liste d'objets/dictionnaires (très utilisé en développement web)
# indent=4 permet de rendre le fichier texte JSON lisible à l'œil humain (sauts de lignes et tabulations)
fichier_json = "donnees_exportees.json"
df_simple.to_json(fichier_json, orient='records', indent=4, force_ascii=False)
print(f" Le DataFrame a été exporté au format JSON sous le nom : '{fichier_json}'")
```
